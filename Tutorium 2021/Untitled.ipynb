```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def fit_func(x, a,b):
    # Curve fitting function
    return a * x**2 + b*x
10/93.
l=np.array([0,42.05*78/93,47,91, 116, 260])*10./78
v=np.array([0,6.25,12.5,25,37.5,50])
dv=np.array([0.12,0.18,0.10,0.21,0.12,1.6])*2
vf=np.linspace(0,50,100)

params = curve_fit(fit_func, v, l)
[a,b] = params[0]
#lf = np.poly1d(np.polyfit(v, l, 2))

plt.errorbar(v, l, yerr=dv, fmt='ok');
plt.plot(vf,a*vf**2+b*vf,'-')
plt.xlabel('Piezo Amplitude (V)')
plt.ylabel('Displacement ($\mu$m)');
;
```
