## 2. Matrixmethoden der paraxialen Optik


Wir wollen nun ein Verfahren entwickeln, mit dem wir Höhe und dem Steigungswinkel des Strahles an jeder Stelle des optischen Systems berechnen können. Für gegebene Eingangsdaten $h_E$, und $\sigma_E$ beim Achsenpunkt $T$ wollen wir die Ausgangsdaten $h'_A$ und $\sigma'_A$ im Punkt $T′$ ermitteln.

### 2.1 Translationsmatrix

Translation eines Strahles in einem homogenen Medium. Betrachten wir den Strahl unten von Punkt 0 nach Punkt1: nach einer Länge $l$, hat sich der Winkel $\sigma$ nicht verändert $\sigma_1=\sigma_0$ und die Höhe $h_1$ ist $h_1=h_0-l\,\tan\sigma_0$. Achten Sie auf die Vorzeichen!



<img src="figures/2.1_1.png" alt="1.1_1.png" style="width: 500px;"/>

Mit Hilfe der paraxialen Näherung erhalten Sie für $\tan\sigma_0\approx\sigma_0$

$$h_1=h_0 \cdot 1 + \sigma_0\cdot(-l)$$
$$\sigma_1=h_0\cdot 0 +\sigma_0\cdot 1$$

oder in Matrixschreibweise

$$
\left(
\begin{matrix}
h_1\\
\sigma_1
\end{matrix}
\right) = 
\left(
\begin{matrix}
1 & -l\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
h_0\\
\sigma_0
\end{matrix}
\right)
$$

Damit ist die Translationsmatrix

$$
\mathrm{T}=
\left(
\begin{matrix}
1 & -l\\
0 & 1
\end{matrix}
\right)\quad .
$$

### 2.2 Brechungsmatrix für Kugelfläche

Bei der Brechung des Strahles an einer Kugelfläche bleibt die Höhe des Strahles erhalten (keine Translation). Für die Winkeländerung benutzten wir unsere zuvor hergeleiteten Gleichungen.

<img src="figures/2.2_1.png" alt="2.2_1.png" style="width: 500px;"/>

Für die Winkel gilt $\sigma'=\varphi-\epsilon'=\frac{h}{r}-\epsilon'$ und $\sigma=\varphi-\epsilon=\frac{h}{r}-\epsilon$.

Mit dem Brechungsgestz in paraxialer Form $n\,\epsilon=n'\,\epsilon'$ erhalten wir:

$$\sigma'=-\frac{n}{n'}\epsilon + \frac{h}{r}=\left( \frac{n'-n}{n'\,r}\right) h + \frac{n}{n'}\sigma$$

Wir können wiederum die zwei linearen Gleichungen für $\sigma'=\sigma'(h,\sigma)$ und $h'=h'(h,\sigma)$ aufstellen:

$$h'=h\cdot1+\sigma\cdot 0$$
$$\sigma'=h\left(\frac{n'-n}{n'\,r}\right) + \sigma\,\frac{n}{n'}$$

Oder in Matrxiform 

$$
\left(
\begin{matrix}
h'\\
\sigma'
\end{matrix}
\right) = 
\left(
\begin{matrix}
1 & 0\\
\frac{n'-n}{n'\,r} & \frac{n}{n'}
\end{matrix}
\right)
\left(
\begin{matrix}
h\\
\sigma
\end{matrix}
\right)
$$

Die Brechungsmatrix für die Kugelfläche ist

$$
\mathrm{B}=
\left(
\begin{matrix}
1 & 0\\
\frac{n'-n}{n'\,r} & \frac{n}{n'}
\end{matrix}
\right)\quad .
$$

### 2.3 Matrizen für dicke und dünne Linsen

Wir können jetzt eine Matrix ableiten, die eine dicke Linse beschreibt. Das Beispiel in der Abbildung ist eine positive Meniskuslinse:

<img src="figures/2.3_1.png" alt="2.3_1.png" style="width: 500px;"/>

Diese besteht aus einer Brechungsmatrix $\mathrm{B_1}$ für die erste Kugelfläche, einer Translationsmatrix $\mathrm{T_1}$ und einer zweiten Brechungsmatrix $\mathrm{B_2}$ für die zweite Kugelfläche.
$$
\left(
\begin{matrix}
h_3\\
\sigma_3
\end{matrix}
\right)
=\mathrm{B_2}\,\mathrm{T_1}\,\mathrm{B_1}\left(
\begin{matrix}
h_0\\
\sigma_0
\end{matrix}
\right)
$$


Die Strahlmatrix $\mathrm{M_{dick}}$ einer dicken Linse ist damit

$$\mathrm{M_{dick}}=
\left(
\begin{matrix}
1 & 0\\
\frac{n-n_L}{n\,r_2} & \frac{n_L}{n}
\end{matrix}
\right)
\left(
\begin{matrix}
1 & -d\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
\frac{n_L-n}{n_L\,r_1} & \frac{n}{n_L}
\end{matrix}
\right)\tag{2.3.1}
$$

Die dünne Linse $d=0$ in Luft $n=n'$ vereinfacht sich zu 

$$\mathrm{L}=
\left(
\begin{matrix}
1 & 0\\
\frac{n-n_L}{n\,r_2} & \frac{n_L}{n}
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
\frac{n_L-n}{n_L\,r_1} & \frac{n}{n_L}
\end{matrix}
\right)=
\left(
\begin{matrix}
1 & 0\\
\frac{n_L-n}{n}\left(\frac{1}{r_1}-\frac{1}{r_2}\right) & 1
\end{matrix}
\right)=\left(
\begin{matrix}
1 & 0\\
\frac{1}{f'} &1
\end{matrix}
\right)
\tag{2.3.2}
$$

In der letzten Gl. (2.3.2) haben wir den Ausdruck der bildseitigen Brennweite $f'$ einer dünnen Linse eingesetzt, Gl. (1.2.8).

### 2.4 Systemmatrix

Wir können obige Ableitung für eine dicke Linse verallgemeinern und für ein beliebiges paraxiales optische System die Strahlgleichung als Matrizenmultiplikation schreiben.


$$\left(
\begin{matrix}
h'\\
\sigma'
\end{matrix}
\right)
=\mathrm{M_K}\,\mathrm{M_{K-1}}\,\dots\,\mathrm{M_1}\left(
\begin{matrix}
h_E\\
\sigma_E
\end{matrix}
\right)
$$

Die Matrix $M=\mathrm{M_K}\,\mathrm{M_{K-1}}\,\dots\,\mathrm{M_1}$ nennt man Systemmatrix.

$$\mathrm{M}=
\left(
\begin{matrix}
A & B \\
C & D
\end{matrix}
\right)$$

Die Gleichungen für die Höhe $h'_A$ und den Winkel $\sigma'_A$ der Ausgangsebene sind

\begin{eqnarray*}
h'_A & =& A\,h_E + B\,\sigma_E\\
\sigma'_A & =& C\,h_E + D\,\sigma_E\tag{2.4.1}
\end{eqnarray*}

Wir können nun Spezialfälle der Systemmatrix betrachten, bei denen eines der Matrixelement Null ist.

#### 2.4.1 Matrixelement $D=0$

<img src="figures/2.4_1.png" alt="2.4_1.png" style="width: 400px;"/>

Dann haben alle Strahlen die von der der Gleichen Höhe der Eingangsebene kommen den gleichen Winkel in der Ausgangsebene. D.h. die Strahlen aus dem Punkt $h_E$ werden parallel abgebildet. Die Eingangsebene ist daher die objektseitige Brennebene.

#### 2.4.2 Matrixelement $A=0$

<img src="figures/2.4_2.png" alt="2.4_2.png" style="width: 400px;"/>

Alle Strahlen die unter dem gleichen Winkel $\sigma_E$ aus der Eingangsebene kommen, werden auf die gleiche Höhe $h'_A$ fokussiert. Damit ist die Ausgangsebene die bildseitige Brennweite.

#### 2.4.3 Matrixelement $B=0$

<img src="figures/2.4_3.png" alt="2.4_3.png" style="width: 400px;"/>

Die Strahlen die unter einem beliebigen Winkel $\sigma_E$ von der Eingangsebene aus der Höhe $h_E$ kommen werden alle auf eine Höhne $h'_A$ fokussiert. Damit sind diese beiden Höhen konjugierte Punkte. Der Abbildungsmaßstab ist daher $\beta'=\frac{h_A'}{h_E}=A$.

#### 2.4.4 Matrixelement $C=0$

<img src="figures/2.4_4.png" alt="2.4_4.png" style="width: 400px;"/>

Parallele Strahlen mit Winkel $\sigma_E$ verlassen das optische System unter dem Winkel $\sigma'_A$. Damit gibt das Matrixelement $C$ die bildseitige Winkelvergrößerung der optischen Systems.


<b>Beispiel: Bestimmung der Systemmatrix einer dicken Linse</b>

[Positive Meniskus Linse #LE1076](https://www.thorlabs.com/newgrouppage9.cfm?objectgroup_id=130) mit  $r_1=30.3\,$mm, $r_2=65.8\,$mm und Mittendicke $d=9.7\,$mm, BK7 Glas $n_L=1.517$, ausserhalb ist Luft $n=1.0$.

<img src="figures/example_lens.jpg" alt="example_lens.jpg" style="width: 400px;"/>

Das Programm unten benutzt die Systemmatrix and die Bedingung $A=0$ für ein paralleles Strahlenbündel um die bildseitige Brennweite zu finden. Dazu wird die Systemmatrix von Gl. (2.3.1) um eine zweite Translation $\mathrm{T_2}$ erweitert. Nun ist $\mathrm{M = T_2 B_2 T_1 B_1}$:


$$\mathrm{M}=
\left(
\begin{matrix}
1 & -z'\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
\frac{n'-n_L}{n'\,r_2} & \frac{n_L}{n'}
\end{matrix}
\right)
\left(
\begin{matrix}
1 & -d\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
\frac{n_L-n}{n_L\,r_1} & \frac{n}{n_L='}
\end{matrix}
\right)\tag{2.3.1}
$$


Wenn in der Matrix $\mathrm{M}$ das Element $A=0$ ist, dann haben wir die bildseitige Brennweite $z'$ erreicht.
Das Programm bestimmt nun den Nullpunkt des Elements $A$ in der Unterroutine ```A(zp)``` mit der importierten Funktion ```fsolve()```. Diese benutzt einen iterativen Löser um sich der Nullstelle von ```A(zp)``` mit dem Startwert ```startzp``` zu finden.

Als Ergebnis erhalten Sie $f'=88.53\,$mm. Das Datenblatt spricht von $88.8\,$mm.

````python
import numpy as np
from scipy.optimize import fsolve

#Thorlabs Lens LE1076
#all values from catalogue and in mm
nL=1.517 #BK7 glass at 580nm
n=1.0
r1=30.3
r2=65.8
d=9.7 

B1 = np.matrix([[1,            0   ],
                [(nL-n)/nL/r1, n/nL]])
T1 = np.matrix([[1, -d],
                [0, 1]])
B2 = np.matrix([[1,           0   ], 
                [(n-nL)/n/r2, nL/n]])

startzp=10. #geschaetzer Nullpunkt

#Berechnung von A als Funktion von z'
def A(zp):
    M = np.matrix([[1, -zp],[0, 1]]) * B2 * T1 * B1
    return M[0,0].flatten()

#Bestimmung des Nullpubnktes der Funktion A(z), dann ist z'=f'.
fp=fsolve(A, startzp)[0] 

print("Bildseitige Brennweite f' ={:7.2f}mm".format(fp))
````

##### 2.5 Kardinalpunkte der optischen Systems am Beispiel der dicken Linse

Die sechs Kardinalpunkte helfen uns den paraxialen Strahlengang eines optischen Systems zu zeichnen. Diese sechs Punkte auf der optischen Achse sind die

* Objektbrennpunkt ($\overline{F}$) und Bildbrennpunkt ($F'$) 
* die konjungierten Hauptpunkte $H$ und $H'$
* die konjungierten Knontenpunkte $K$ und $K'$

Bei der dicken Linse läuft der Strahl ausgehend vom Objektbrennpunkt $\overline{F}$ parallel zur optischen Achse. Ein einfallender Strahl parallel zur optischen Achse wird durch die Linse gebrochen und läuft durch den bildseitigen Brennpunkt. Die ein/auslaufenden und aus/einlaufenden Strahlen treffen sich an den Hauptebenen der Linse. Bei der dünnen Linse ist das die Mitte der Linse, bei dicken Linsen sind das die beiden Hauptebenen die durch die Hauptpunkte $H$ und $H'$ gehen, siehe Abbildung:

<img src="figures/2.5_1.png" alt="2.5_1.png" style="width: 600px;"/>

Bei der dünnen Linse geht der Mittelpunktstrahl ungebrochen hindurch, bei der dicken Linse gibt es auch einen Strahl nicht seinen Winkel ändert, es gibt aber eine Parallelverschiebung des Strahls. Dieser Punkt ist der objektseitige Knotenpunkt $K$, der auslaufende Strahl läuft parallel dazu und duch den bildseitigen Knotenpunkt $K'$.

<img src="figures/2.5_2.png" alt="2.5_2.png" style="width: 300px;"/>

In der dünnen Linse fallen die beiden Hauptpunkte und die Knotenpunkte auf die Mitte der Linse zusammen. Bei dicken Linsen und allen anderen optischen Systeme liegen diese Punkte nicht notwendgerweise zwischen der Eingangsebene und Ausgangsebene. 


Die Lage der Kardinalpunkte können aus der Systemmatrix gewonnen werden. Die Abbildung unten zeigt die Größen, die berechnet werden können. Die Herleitung der Gleichungen finden Sie im Lehrbuch (Pedrotti *et al.*).

<img src="figures/2.5_3.png" alt="2.5_3.png" style="width: 550px;"/>

|Strecken &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;     |Name |
|---|---|
|$\overline{f}_E=-\frac{D}{C}$ |objektseitiger Abstand des Brennpunktes $\overline{F}$|
|$f_{A'}=\frac{A}{C}$ |bildseitiger Abstand des Brennpunktes $F'$|
|$s_{EH}=\frac{n_E/n'_A-D}{C}$|objektseitiger Abstand des Hauptpunktes H|
|$s_{AH'}=\frac{A-1}{C}$|bildseitiger Abstand des Hauptpunktes H'|
|$s_{EK}=\frac{1-D}{C}$|objektseitiger Abstand des Knotenpunktes K|
|$s_{AK'}=\frac{A-n_E/n'_A}{C}$|bildseitiger Abstand des Knotenpunktes K'|
|$\overline{f}=\overline{f}_E-s_{EH}=-\frac{n_E/n'_A}{C}$|objektseitige Brennweite relativ obj. Hauptebene|
|$f'=f'_A-s'_{AH'}=\frac{1}{C}$|bildseitige Brennweite relativ bild. Hauptebene|

<b>Beispiel: Systemmatrix eines zweilinsigen optischen Systems</b>

Wir berechnen die Hauptpunkte eines optischen Systems mit zwei dünnen Linsen mit Brennweiten $f'_1$ und $f'_2$ die im Abstand $e$ angeordnet sind:

<img src="figures/2.5_4.png" alt="2.5_4.png" style="width: 400px;"/>

Die Systemmatrix ist

$$\mathrm{M}=
\left(
\begin{matrix}
1 & 0\\
\frac{1}{f'_2} & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & -e\\
0 & 1
\end{matrix}
\right)
\left(
\begin{matrix}
1 & 0\\
\frac{1}{f'_1} & 1
\end{matrix}
\right)=
\left(
\begin{matrix}
1-\frac{e}{f'_1} & -e\\
\frac{1}{f'_2}\left( 1-\frac{e}{f'_1} \right)+\frac{1}{f'_1} & 1-\frac{e}{f'_2}\\
\end{matrix}
\right)
$$

Die bildseitige Brennweite ist $f'=\frac{1}{C}$:

$$\frac{1}{f'}=\frac{1}{f'_1}+\frac{1}{f'_2}-\frac{e}{f'_1\,f'_2}\tag{2.5.1}$$

Weiterhin sind

$$s_{EH}=s_{EK}=\frac{f'}{f'_2}e$$

und

$$s_{AH'}=s_{AK'}=-\frac{f'}{f'_1}e\quad .$$
