# 8. Holographie

## 8.1 Geschichte

Bereits 1948 hat Dennis Gabor eine Methode zur dreidimensionalen Photographie von Objekten vorgestellt, die *Holographie*. Dazu benötigt man kohärentes Licht, das erst mit der Erfindung des Laser zur Verfügung stand. Holografie wird in der holgrafischen Inteferometrie benutzt wenn Objekte das Licht diffus streuen und die Objekte zu einer späteren Zeit mit sich selbst interferometrisch verglichen werden können. Auch können mit modernen Kameras Hologramme digital aufgezeichnet und ausgewertet werden. Die [digitale holografische Mikroskopie](https://en.wikipedia.org/wiki/Digital_holographic_microscopy) (DHM) ist ein weiters wichtiges Anwendungsfeld.

## 8.2 Aufnahme eines Hologramms

In der Photographie wird die Intensität des Lichtes aufgezeichnet. Dazu wird gewöhnlicherweise das Objekt mit inkohärentem Licht beleuchtet. Durch diese Aufnahmetechnik verliert man die Phaseninformtaion des Lichtes. In der Holographie wird die Amplitude und Phase des Lichts aufgezeichnet, indem zusaätzlich zum vom Objekt abgestrahltne Licht eine Referenzwelle addiert wird. Dazu wird ein kohärente Lichtquelle benötigt. Die Phaseninformation wird in dem Interferenzmuster abgespeichert. 

Die Rekonstruktion des holographischen Bildes geschieht, indem das entwickelte Hologramm mit der Referenzwelle wiederum beleuchtet wird.  

Nun wollen wir die Aufzeichnung des Hologramms uns genauer anschauen:

<img src="figures/8.2_1.png" alt="8.2_1.jpg" style="width: 450px;"/>

Wir nennen die Welle die vom Objekt abgestrahlt wird $S(x,y,t)$ und die Referenzwelle, die direkt auf das Hologramm fällt $R(x,y,t)$. Auf dem Hologramm $z=0$ fällt dann das elektrische Feld

$$E(x,y,t)=S(x,y,t) + R(x,y,t)\tag{8.2.1}$$

Nehmen wir monochromatisches Licht an ist $R(x,y,t)=R(x,y)\,e^{-i\omega t}$ und das ruhende Objekt $S(x,y,t)=S(x,y)\,e^{-i\omega t}$

Damit erhalten wir für die Intensitätsverteilung auf der Hologrammplatte


\begin{eqnarray*}
I(x,y)&=&\left|R(x,y)+S(x,y)\right|^2=(R+s)(R+S)^*\\
&=&RR^*+SS^*+R^*S+RS^*\tag{8.2.2}
\end{eqnarray*}


Der photographische Film ist empfindlich für die Gesamtenergie pro Fläche die während der Belichtungszeit $t_B$. Der Film wandelt diese Größe in eine optische Dichte und Änderung des Brechungsindexes um. Belichtung $B$ is

$$ B(x,y)=\int_0^{t_B} I(x,y,t)\,\mathrm{d}t=\int_0^{t_B} E(x,y,t)E^*(x,y,t)\,\mathrm{d}t$$

Dadurch bildet der Film eine komplexe Durchlässigkeit für die Amplitude des elektrischen Feldes

$$\tau=\frac{E_t}{E_e}=T(x,y)\,e^{i\varphi(x,y)}$$

<img src="figures/8.2_2.png" alt="8.2_2.jpg" style="width: 200px;"/>

Es gibt zwei Grenzfälle, nämlich das Amplitudenhologramm ($\varphi=$konst.) und das Phasenhologramm ($\T=$konst.). Das Phasenhologramm hat den Vorteil, dass es fast kein Licht absorbiert und daher hellere Bilder bei der Rekonstruktion erzeugt.

Bei der Standardnelichtung und Entwicklung von holographischen Filmen wird ein Amplitudenhologramm erzeugt. Die Abbildung unten zeigt eine typische Transmission $T$ eines entwickelten Films in Abhängigkeit von der Belichtung $B$

<img src="figures/8.2_3.png" alt="8.2_3.png" style="width: 300px;"/>

und kann in einem mittleren Bereich um die Belichtung $B_A$ linear entwickelt werden:

$$T=a-b\,B=a-b\,I\,t_B$$

Damit wird die Transmission der Amplitudenholograms

$$T(x,y)=a-b\,t_B\left(R R^* + S S^* + R^* S +R S^*\right)(x,y)\quad.\tag{8.2.3}$$

Phasenholograme ist das ganz ähnlich, wir nehmen wierderum einen Arbeitspunkt der Belichtung $B'_A$ an und entwickeln die Phase

<img src="figures/8.2_4.png" alt="8.2_4.png" style="width: 300px;"/>


$$\varphi(I)=a'+b'\,t_B\,I$$

Die Transmission des Films ist nun


$$\tau=e^{i\varphi(I)}$$

Wenn die Phase $\varphi(I)\ll\pi/2$ können wir die Exponentialfunktion linearisieren

$$\tau=e^{i\varphi(I)}\approx1+i\varphi(I)=(1+ia')+
ib't_B\left(R R^* + S S^* + R^* S +R S^*\right)\tag{8.2.4}$$. 
Gl. (8.2.3) für das Amplitudenhologramm und Gl. (8.2.4) für das Phasenhologram sind sehr ähnlich bis auf die unterchiedlichen Konstanten.

## 8.3 Wiedergabe/Rekonstruktion eines Hologramms

Bei der Wiedergabe (Rekonstruktion) des Hologramms wird das Hologram mit der gleichen Referenzwelle beleuchtet, die auch bei der Aufnahme verwendet wurde. Dadurch wird die Referenzwelle $R(x,y)e^{i\omega t}$ mit der komplexen Transmission moduliert $\tau(x,y)$ moduliert. Ist das Hologram ein Amplitudenhologram erhalten wir

\begin{eqnarray*}
E_t(x,y)&=&T(x,y)\,E_e=T(x,y)\,R\\
&=&R\,a - b\,t_B\,R\left(R R^* + S S^* + R^* S +R S^*\right)\quad .\tag{8.3.1}
\end{eqnarray*}

Wir erhalten 4 Terme die folgende Bedeutung haben

1. $(a-b\,t_B|R|^2)R\propto R$ ist die Referenzwelle $R$ multipliziert mit einem konstanten Faktor (nullte Beugungsordnung)

2. $b\,t_B\,|S|^2\,R$ ist die um eine Speckelmuster $|S|^2$ verbreiterte nullte Beugungsordnung

3. $b\,t_B\,|R^2|S\propto S$ direkte Bild (virtuell)

4. $b\,t_B R^2\,S^*$ konjugiertes Bild (reell)


### 8.3.1 Lage der Bilder

Die Lage der Bilder sind in der unteren Abbildung angezeigt. Das direkte Bild ist ein virtuelles Bild, dass hinter der Hologramm liegt ähnlich der Lage einse Objektes hinter einem Fenster. Das konjugierte Bild ist reell, liegt vor der Hologramm und kann mit einem Schirm angeschaut werden.
<img src="figures/8.3_1.png" alt="8.3_1.png" style="width: 500px;"/>
Nehmen wir einfachshalber an, dass das Objekt eine ebene Welle ist, die auf das Hologramm unter dem Winkel $\gamma$ trifft. 
Die Richtung der konjugierten Welle $S^*$ ist dann mit

$$S=e^{ikx\sin \gamma}\rightarrow S^*=e^{-ikx\sin\gamma}=e^{ik\sin(-\gamma)}\quad .\tag{8.3.2}$$

Somit kann man die Welle $S^*$ als eine Welle, die das Hologramm unter einem Winkel $-\gamma$ verläßt, auffassen.
Jetzt nehmen wir wiederum $R$ als ebene Welle $R=\exp(ikx\sin\alpha)$, die auf die Hologrammplatte unter dem Winkel $\alpha$ auftritt. Wegen $R^2$ haben wir

$$R^2=e^{ikx2\sin\alpha}=e^{ikx\sin\alpha'}\quad .\tag{8.3.3}$$

Für kleine Winkel ist $\alpha'=2\alpha$. Dies haben wir auch in der Abbildung oben angenommmen. Aus den Gln. (8.3.2) und (8.3.3) folgt für das Bild $R^2S^*$

\begin{eqnarray*}
R^2\,S^*&=&e^{ikx2\sin\alpha}\,e^{i k x \sin(-\gamma)}\\
&=&e^{ikx(2\sin\alpha+\sin(-\gamma))}
\end{eqnarray*}

Nehmen wir nun wieder an, dass das konjugierte Bild eine ebene Welle ist, die das Hologramm unter dem Winkel $\beta$ verlâßt, d.h. $R^2\,S^*=\exp(ikx\sin\beta)$ erhalten wir die Beziehung

$$\sin\beta=2\sin\alpha+\sin(-\gamma)\quad .$$

Mit größer werdendem $\alpha$ oder $\gamma$ verformt sich das konjugierte Bild immer mehr bis

$$\sin\beta=2\sin\alpha+\sin(-\gamma)>1$$

und der Winkel $\beta$ nicht mehr existiert. Dann verschwindet dieser Teil des konjugierten Bildes. Bei senkrechtem EInfall der Referenzstrahls vereinfachen sich die Lagen:

## 8.5 Digitales Hologramm

Ein Hologramm kann statt mit einem optischen Aufbau auch mit einem Algorithmus berechnet werden. Danach gibt man es auf ein geeignetes Medium aus. Das könnte eine photographisch Belichtete Maske sein, die die notwendige Auflösung hat. Man kann auch Flüssigkristallanzeigen als Phasenhologramme benutzten. 

### 8.5.1 Numerische Simulation eines Hologramms

Nehmen wir zur Einfachheit an, dass ein Objekt aus einzelnen Punkten $P_l,\,l=1,2,\dots\,L$ an den Positionen $(x_l,y_l,z_l)$. Die Hologrammplatte liegt in der Ebene $z=0$ und besteht aus den Bildpunkten $\{(x_m,y_n,0),\,m=1,\dots,\,M,\,n=1,\,\dots,\,N\}$, siehe auch die untenstehende Abbildung 

<img src="figures/8.5_1.png" alt="8.5_1.png" style="width: 450px;"/>

Jeder Punkt $P_l$ ist Quelle einer sphärische Welle die zu einer komplexen Amplitude an der Stelle $(x_m,y_n)$ auf der Hologrammplatte führt.

$$e_{lmn}=A_l\,e^{i\varphi_l}\,\frac{i\,k\,r_{lmn}}{r_{lmn}}$$

$A_l$ und $\varphi_l$ sind die Amplitude und Phase der sphärischen Welle vom Punkt $P_l$. $r_{mnl}$ ist der Abstand des Punktes vom Hologrammpunkt $(x_m,y_n)$ 

$$r_{lmn}=\sqrt{\left(x_l-x_m\right)^2+\left(y_l-y_n\right)^2+z_l^2}$$

Die Amplitude der Objektwelle sind die summierten Wellen

$$E_{Smn}=\sum_{l=1}^L\,E_{lmn}\quad .$$

Zusammnen mit der Referenzwelle $E_R$ erhalten wir als Intensität am Punkt $(x_m,y_n)$ 

$$I_{mn}=\left|E_{Rmn}+E_{Smn}\right|^2=|E_{mn}|^2\quad \tag{8.5.1}.$$

Prinzipiell kann diese Intensität an jedem Punkt auf der Hologrammplatte berechnet werden. Der Rechenaufwand ist aber enorm, wenn wir hochauflösende Hologramme von komplexen Geometrien berechnen wollen. Wir machen ein paar Vereinfachungen:

1. Jeder Objektpunkt emittiert eine Lichtwelle mit der gleichen Amplitude $A_l=\mathrm{konst.}$ für alle $l$. 

2. Jede Objektwelle emittiert mit der gleichen Phase $\varphi_l=0$ für alle $l$.

3. Wenn das Objekt weit vom Hologramm entfernt und klein ist können wir die Amplitude des der Objektwellen gleich setzten, nämlich
  
  $$\frac{A_l\,e^{i\varphi_l}}{r_{mnl}}=\frac{A_l}{r_{mnl}}=\mathrm{konst.}=1\quad,$$

  wobei die 1 hier frei gewählt ist.

4. Jetzt fehlt noch der Ausdruck $E_{lmn}=e^{i k r_{lmn}}$, und wählen als Wellenzahl $k=1$

  $$E_{lmn}=e^{i\,r_{lmn}}$$
  
5. Die Referenzwelle fällt senkrecht auf die Hologrammplatte. Wiederum wählen wir eine konstante Phase so dass gilt:

  $$E_{Rmn}=E_R=\mathrm{konst.},\quad\mathrm{für\, alle\,}m,n; E_R\,\mathrm{real}$$ 


Wir können die Berechnung und Herstellung der Hologramme aber noch weiter vereinfachen wenn anstatt von Graustufen die aus Gl. (8.5.1) folgen, ein Schwarz-Weiss Muster genommen wird. In der Abbildung wurde ein mittlerer Helligkeitswert $\overline{I}$ benutzt; alles darüber ist schwarz, darunter ist hell. 

<img src="figures/8.5_2.png" alt="8.5_2.png" style="width: 450px;"/>

Betrachten wir nun die Summation der Objektwellen $E_{lmn}$ in der komplexen Ebene. Wenn die Intensität der einzelnen Objektwellen viel kleiner als die Intensität der Referenzwelle ist, ist die mittlere Intensität vorangig durch

$$\overline{I}\approx|E_R|^2=E_R^2$$

Wir können uns die Addition der Objektwellen mit der Referenzwelle graphisch vorstellen:

<img src="figures/8.5_3.png" alt="8.5_3.png" style="width: 550px;"/>

Wenn der Realteil der Summe der Intensitäten der Objektwelle $>0$ ist, dann erhalten wir einen schwarzen Punk, wenn dem nicht so ist, weiss. 

Damit lässt sich ein einfaches Kriterium geben, wie ein Punkt auf der Hologrammplatte eingefärbt wird:

\begin{eqnarray*}
Re(E_{Smnn})>0 &\rightarrow & \mathrm{black\,dot,}\\
\mathrm{else} &\rightarrow & \mathrm{none.}
\end{eqnarray*}

Als Konsequenz benötigen wir nun nicht mehr die Referenzwelle und müssen nur den Realteil der summierten Objektwelle beachten der damit nur vom $\cos$ des Abstandes $r_{lmn}$ abhängt:


$$
Re(E_{Smn})=\sum_{i=1}^L\cos(r_{lmn})
$$


Unten finden Sie ein Programm, mit dem Sie ein digitales (binäres) Hologramm von einer Punktquelle herstellen können. Das können Sie für die Übung verwenden.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
nx=480
ny=320
#Points
P=np.array([[nx/2, ny/2, -800]])

npts=len(P)
H=np.zeros((ny,nx))
dyv=np.zeros(npts)

for j in range(ny):
    
    for i in range(npts):
        dyv[i]=(P[i][1]-j)**2
    
    for k in range(nx):
        sum=0.
        for i in range(npts):
            dx=P[i,0]-k
            sum += np.cos((dx*dx+dyv[i]+P[i,2]*P[i,2])**.5)
        
        if (sum>=0):
            H[j,k]=1.
            
plt.figure(figsize=(20,20))
plt.imshow(H);
```
