### 12.1 Nichtlineare Medien und Frequenzverdopplung

In einem linearen isotropen Medium ist die Polarisation $P$ unabhängig von der elektrischen Feldstärke

$$\vec{P}=\epsilon_0\,\chi\vec{E}\quad .\tag{12.1.1}$$

Hier ist die elektrische Suszeptibilität $\chi$ als Konstant angenommen. Diese Annahme wird bei genügend hohen Feldstärken ungültig und die Suszeptibilität wird eine Funktion von der Feldstärke, $\chi=\chi(E)$.  Wir können $\chi$ entwickeln und schreiben

$$\chi(E)=\chi_1+\chi_2\, E+\chi_2\,E^2+\chi_3\,E^3+\,\dots$$
so daß

$$P=\epsilon_0\left(\chi_1+\chi_2\, E+\chi_2\,E^2+\chi_3\,E^3+\,\dots\right)\quad . \tag{12.1.2}$$

Der erste Term in Gl. (12.1.2) umfasst die lineare Optik. Die höheren Terme beeinflussen die Polarisation nur bei sehr starken $E$-Feldern.

$$P=\underbrace{P_1}_{\mbox{linear}}\,+\, \underbrace{P_2 \, P_3 \,+\, \dots}_{\mbox{nichtlineare Terme}}$$

Die Frequenzverdopplung ist ein nichtlinearer Effekt der aus dem Term 2. Ordnung resultiert:

$$P_s=\epsilon_0\,\chi\,E^2$$

Frequenzverdopplung findet nur in nicht nicht homogenen Stoffen oder Kristallen ohne Punktsymmetrie auf, also nicht in Flüssigkeiten.

Es kann leicht gesehen werden, dass ein quadratischer Term in $E$ zur Frequenzverdopplung führt wenn wir ein $E$-Feld so schreiben:

$$E=\hat{E}\,\cos \omega t$$

Wegen $\cos^2\alpha=1/2(1+\cos 2\alpha)$ ist 
$$P_2=\epsilon_0\,\chi_2\,\hat{E}^2\cos\omega t=\frac{1}{2}\epsilon_0\,\chi_2\,\hat{E}^2+\frac{1}{2}\epsilon_0\,\chi_2\,\hat{E}^2\cos 2\omega t\quad\tag{12.1.3} .$$

Es werden also bei genügend hohen Feldstärken in einem nichtlinearen Medium das Licht in der Frequenz verdoppelt, d.h. die Lichtwellenlänge halbiert sich. Die Konversion der fundamentalen Frequenz $\omega$ auf $2\,\omega$ ist dann effektiv, wenn die Ausbreitungsgeschwindigkeiten $c_\omega$ für $\omega$ und $c_{2\omega}$ für $2\,\omega$ identisch sind. Normale Dispersion im Medium verhindert dies, und $c_\omega > c_{2\omega}$. Als Ausweg bietet sich die <em>Phasenanpassung</em> beider Strahlen an. Dazu verwendet man einen doppelbrechenden Kristall in dem die Grundwelle als ordentlichen Strahl mit der gleichen Geschwindigkeit propagiert wie die Oberwelle ($\2\,\omega$) als außerordentlicher Strahl. Durch Orientierung des Kristalls und/oder Änderung der Temperatur kann die Phasengeschwindigkeit des außerordentlichen Strahls auf den ordentlichen Strahl angepasst werden.

### 12.2 Kerr Effekt

<img src="figures/12.2_1.png" style="width: 450px"> 

In isotropen Medien wie Flüssigkeiten oder Gläser treten keine Effekte 2. Ordnung auf. Die Polarisation wird hier durch einen Term 3. Ordnung $P_e\propto E^3$ modifiziert. Ein elektrisches Feld kann die Moleküle senkrecht zur Ausbreitungsrichtung der Lichtstrahls ausrichten. Dabei wird das Medium doppelbrechend. Dann kann ein Lichtstrahl der $45^\circ$ zum elektrischen Feld polarisiert (Brechzahl $n_o$) in eine außerordentlichen Strahl mit Brechzahl $n_{ao}$ gleicher Amplitude aufspalten. Beide Strahlen sind bei einer Phasendrehung von $\delta =\pi$ zueinander senkrecht polarisiert.

Beim elektrischen Kerr Effekt, d.h. das statische elektrische Feld multipliziert mit dem Wechselfeld des Lichtstrahls, kann die Brechzahldifferenz vereinfacht beschrieben werden:

$$\Delta n=n_{ao}-n_o=K\,E^2\,l$$

und die Phasenverschiebung

$$\delta=\frac{2\pi K U^2 l}{d^2}\quad ,$$

wobei $U$ die Spannung an der Zelle ist, die anderen Konstanten sind in der Abbildung oben definiert. Setzen wir $\delta=\pi$, also eine $\lambda/2$ Platte, dann erhalten Sie für die notwendige Spannung 

$$U_{\lambda/2}=\frac{d}{\sqrt{2\,K\,l}}\quad .$$


Die Konstante $K$ ist nur für wenige Flüssigkeiten hoch, z.B. Nitrobenzol $K=2.4\,10^{-12}\,$mV$^{-2}$ oder Schwefelkohlenstoff $K=3.6\,10^{-14}\,$mV$^{-2}$. Dadurch benoötigt man hohe Spannungen, um merkbar die Polarisation des außerordentlichen Strahles zu drehen.

Neben dem elektrischen Kerr Effekt gibt es auch den optischen Kerr Effekt. Hier alleine dreht das elektrische Wechselfeld des Lichstrahls die Polarisation des außerordentlichen Strahls. Diese Phänomen erlaubt es Licht mit Licht zu schalten und zum Beispiel Licht im Flug zu fotografieren.

$$n(I)=n+n_2 \, I$$

### 12.3 Akustooptische Modulation
Beim akustooptischen Effekt ändert sich der Brechungsindex durch eine akustische (longitudinale) Welle. Die Ultraschallwellen werden mit einem piezoelektrischen Schallwandler (Transducer) in einer Flüssigkeit oder einem transparenten Festkörper erzeugt.  Die Brechzahl des Mediums vergrößert sich mit dem Druck. Nehmen wir einen linearen Zusammenhang zwischen Druck und Brechnungsindex, dann erzeugt die akustische Welle ein Phasengitter im Medium, an dem die Lichtwelle wie an einem Gitter gebeugt wird. 

Wir wiederholen die Gl. (9.6.1) vom Kapitel 9, die die Intensitätsverteilung am Gitter beschreibt: 

$$I(\nu_x,\nu_y)=|E_0|^2\delta(\nu_y) a^2 \mathrm{sinc}^2(a\,\nu_x)\frac{\sin^2(\pi M d \nu_x)}{\sin^2( \pi d \nu_x)}$$

Maxima treten an Positionen $\pi d \nu_x=\pi d \sin\alpha/\lambda=m\pi$ mit $m=0,\,\pm 1,\,\pm 2,\,\dots$ auf, denn dann ist der Nenner Null (die d'Hospital Regel angewandt verhindert die Singularität).

$$\sin\alpha_{max}=m \frac{\lambda}{d}$$

Jetzt ist der Gitterabstand durch die akustische Wellenlänge $\lambda_s=c_s/f_s$ gegeben, d.h. wir sehen die Maxima des gebeugten Lichtes unter

$$\sin\alpha_{max}=m\frac{\lambda}{\lambda_s}\quad .\tag{12.3.1}$$

Die akustische Welle ist nicht ortsfest, sondern bewegt sich mit der Geschwindigkeit $c_s$ im Medium (z.B. $1500\,m/s$ in Wasser). #### 12.4.1 Lichtausbreitung in Fasern

Wir betrachten die Lichtausbreitung in Fasern mit der geometrischen Optik und Strahlen die in durch die Mittelebene der Faser sich ausbreiten. 
Die Faser besteht aus einem Faserkern mit Brechzahl $n_K$ das von einem Fasermantel mit geringerer Brechzahl $n_M$ umgeben ist. 
 
<img src="figures/12.4_1.png" style="width: 550px"> 

Wir betrachten zuerst Strahl 1, der beim Eintritt in den Mantel (Punkt $A$) gebrochen wird bevor er an Punkt $B$ auf Grenzfläche zwischen $n_K$ und $n_M$ trifft. Ein Teil des Strahles wird reflektiert und der andere Anteil läuft im Medium $n_M$ nachdem dieser vom Lot weg gebrochen wurde. Dieser Vorgang wiederholt sich an Punkten $B$ und $C$. Dabei verliert der Strahl  mehr und mehr an Intensität; der Strahl wird nicht vom Kern geführt. 

Bedingung dafür, dass der Strahl vom Kern geführt wird, ist dass der Winkel $\epsilon$ kleiner als der Winkel für die Totalreflektion $\epsilon_g$ ist.

$$\epsilon_1<\epsilon_g=\arcsin\frac{n_M}{n_K}$$

Um dies zu erreichen, muß der Strahl mit einem kleineren Winkel als bei Strahk in den Kern eingeführt (eingekoppelt) werden. Mit dem Snelliuschen Brechnungsindex erhalten wir den Zusammenhang für Strahl 2

$$n_0\sin u = n_K\sin u'$$

Bei Totalreflektion an Punkt $E$ gilt

$$\sin \epsilon_g=\frac{n_M}{n_K}\tag{12.4.1}$$


Weiterhin ist $\sin u' =\sin (90^\circ-\epsilon_g)=\cos \epsilon_g$.

Wir können nun diesen Akzeptanzwinkel $A_n=n_0\sin u$ mit
$n_K\cos\epsilon_g=n_K\,\sqrt{1-\sin^2 \epsilon_g}$ gleichsetzen.  Daraus folgt mit obiger Gl. (12.4.1)

$$A_n=n_0\sin u=\sqrt{n_K^2-n_M^2}\tag{12.4.2}$$

Dies Gleichung (12.4.2) sagt voraus, dass selbst bei kleinen Brechzahldifferenzen eine großer Öffnungswinkel $2 u$ existiert, unter dem eine Kernmode in die Glasfaser eingekoppelt werden kann, z.B. für eine Quarzglasfaser mit $n_K=1.48$ und $n_M=1.46$ erhalten wir einen Öffnungswinkel $2 u\approx 28^\circ$.

Trotz der Totalreflektion treten beim führen des Strahles Verluste auf. Dies können Streufverluste im Kern sein, Absorption im Kernmaterial oder Unebenheiten am Kern/Mantel Übergang. Die Verluste werden durch das Extinktionsgesetzt quantitfiziet

$$P=P_0\,e^{-K\,l}\quad .$$
Hierbei sind $P_0$ die eintretende Lichtleistung, $P$ die nach Länge $l$ austretende Lichtleistung P und $K$ die Extinktionskonstante.

Generell gibt man die Dämpfung durch die Daämpfungskonstante $\alpha$ in dB pro km an, nämlich

$$\alpha=\frac{10}{l}\log_{10}\frac{P_0}{P}\quad .\tag{12.4.3}$$

Bei einer Faser mit $\alpha=3\,$dB/km verbleiben nach einer Länge von $l=1\,$km nur noch $P/P_0=10^{-\alpha\,l/10}=10^{3/10}\approx=0.5$ also 50% der eingebrachten Leistung übrig.

## 13. Spektrometer

Der Strahlengang beim Transmissionsgitter ist in der Abbildung unten angegeben. Die zu untersuchende Lichtquelle $Q_\lambda$ wird über die Linse $L_1$ auf den Spalt $Sp$ abgebildet. Die Linse $L_2$ kollimiert den Strahl und der fällt auf das Gitter $G$. Die Beugungsordnungen (hier $m=0$ und $m=\pm 1$) werden mit der Linse $L_3$ auf den Schirm $Sc$ abgebildet. 

<img src="figures/13.1_1.png" style="width: 550px"> 

Die Gittergleichung für den Winkel unter dem eine Wellenlänge $\lambda$ in der Beugungsordnung $m$ angebildet wird ist <em> nicht</em> eindeutig!

$$\sin\alpha_{max}=\frac{m\,\lambda}{d}$$

Damit kann und wird es vorkommen, dass unter dem gleichen Winkel $\alpha_{max}$ verschiedene Wellenlängen unter verschiedenen Beugungsordnungen auftreten. Dies schränkt den nutzbarene Spektralbereich ein. Die Abbildung unten zeigt den Winkelbereich für sichtbares Licht $400\,$nm$\,\le\,\lambda\,\le 700\,$nm und verschiedenen Beugungsordnungen eines Gitters mit $d=400$ Linien/mm oder $d=2.5\,\mu$m.

<img src="figures/13.1_2.png" style="width: 550px"> 


Betrachten wir ein Wellenlängenbereich $\lambda_1\,\le\,\lambda\,\le\,\lambda_2$. Die größte nicht überlappende Wellenlänge, $\lambda_2$ in der Ordnung $m$ muß kleiner sein, als die kleinete Wellenlänge $\lambda_1$ der nachfolgenden Ordnung $m+1$. Dieser Wellenlängenbereich $\Delta \lambda=\lambda_2-\lambda_1$ wird nutzbarer oder freier Spektralbereich genannt. Benutzten wir den Wellenlängenbereich von oben erhalten wir

$$\Delta \lambda=\frac{\lambda_1}{m}$$

##### Beispiel 1
Wie groß sind die freien spektrale Bereiche eines Gitters mit der kleinsten Wellenlänge $\lambda=400\,$nm? 

$$m=1\quad \rightarrow\quad \Delta \lambda=\frac{400\,\mathrm{nm}}{1}=400\,\mathrm{nm}$$
$$m=2\quad \rightarrow\quad \Delta \lambda=\frac{400\,\mathrm{nm}}{2}=200\,\mathrm{nm}$$
$$m=3\quad \rightarrow\quad \Delta \lambda=\frac{400\,\mathrm{nm}}{3}=133\,\mathrm{nm}$$


##### Beispiel 2
Wieviele Linien hat das Transmissionsgitter in der Übungsaufgabe? 

Wir bestrahlen das Gitter mit einer bekannten Wellenlänge $\lambda=532\,$nm und messen die Position der ersten Beugung $m=1$. Hier in einem Abstand von $z=3.0\,\pm0.2\,$cm befindet sich das die Beugungsordnung am Ort $x=1.5\,\pm0.2\,$cm. Dazu bestimmen wir zurerst den Winkel $\alpha$ und dann mit der Gittergleichung 

$$d=\frac{m\,\lambda}{\sin \alpha}\approx 840\,\pm 100\, \mbox{Linien pro mm}$$

```python
import numpy as np
z=0.03
x=0.015
m=1.
l=532e-9
alpha=np.arctan(x/z)
d=m*l/np.sin(alpha)
print('{0} Linien pro mm'.format(1e-3/d))
```
