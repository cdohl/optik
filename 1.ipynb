### 1.1.1 Reflexion & Brechung mittels des Huygenschen Prinzips

<b> Reflexion an einem ebenen Spiegel:</b>

<img src="figures/1.1_4.png" alt="1.1_4.png" style="width: 500px;"/>

Konstruktion der Reflexion an einem mit Hilfe des Huygenschen Prinzips. Beim Auftreffen des Strahles mit einer ebenen Wellenfront durch die Punkte $ABC$ werden Elementarwellen ausgesendet. Hier müssen wir berücksichtigen, dass die Wellenfront $DEF$ zu verschiedenen Zeit auf den Spiegel $xy$ aukommt. Wenn wir am Punkt $I$ eine Elementarwelle mit Radius 0 erhalten, hat sich eine Elementarwelle mit Radius $FI - EJ$ vom Punkt $J$ ausgebreitet.

* Der Winkel $\sphericalangle ADx$ ist gleich dem Winkel $\sphericalangle IDG$.
* Das Dreieck $\bigtriangleup DIG$ und $\bigtriangleup DIM$ sind ähnlich.
* Der Winkel $90^\circ - \epsilon$ und $90^\circ -\epsilon'$ sind gleich.

Daher sind der Einfallswinkel zum Lot $PD$, $\epsilon$ und der Ausfallswinkel vom Lot $\epsilon'$ identisch und es gilt Einfallswinkel gleich Ausfallswinkel bei der Reflexion am ebenen Spiegel.


Wie sieht es bei der Reflektion an einem gekrümmten Spiegel aus?


<b>Brechung an einer Ebene:</b>

Oberhalb der Ebene $xy$ breiten sich Wellen mit der Geschwindigkeit $c$ aus und unterhalb mit der Geschwindigkeit $c'$. Wir führen die dimensionslosen Brechnungsindex $n$ ein, so dass

$$c=c_0/n\quad .$$

Natürlich ist dann $c'=c_0/n'$. 

Bei der Brechung betrachten wir die Ausbreitung des Lichtes im Medium unterhalb $xy$. 

<img src="figures/1.1_5.png" alt="1.1_5.png" style="width: 500px;"/>

Die Strecke $FI=c\cdot t$ dann ist auch $DG=c\cdot t$. Die Elementarwellen von $D$, $J$ und von $I$ breiten sich nun nicht mit $c$ sondern mit $c'$ aus. Die Zeit $t$ wird $t=DG/c=n\cdot DG/c_0$.

Während der gleichen Zeit $t$ breitet sich die Elementarwelle von $D$ nur um die Strecke $DM=c'\cdot t$ aus. Die Strecke $DM = c' n DG/c_0=n/n' DG$ lässte sich mit dem Verhältnis der Indices $n/n'$ beschreiben.

\begin{eqnarray*}
\sphericalangle DIF = \epsilon\\
\sphericalangle DIM = \epsilon'\\
\sin\epsilon = \frac{FI}{DI}\\
\sin\epsilon' = \frac{DM}{DI}\\
\frac{\sin \epsilon}{\sin \epsilon'} = \frac{FI}{DM}=\frac{DG}{DM}=\frac{n'}{n}
\end{eqnarray*}

Die letzte Zeile können wir einprägsam schreiben als

$$
n\,\sin\epsilon = n'\,\sin\epsilon'\quad .
$$

Das ist das Snellius'sche Brechungsgesetz.


Aus dem Snellius'sche Brechungsgesetz ergibt automatisch die ein maximaler Einfallswinkel fuer die Brechung, so dass $\epsilon' \ge 90^\circ$ wird:


$$
n\,\sin\epsilon = n'\,\cdot 1 \quad 
$$
⇒

$$\epsilon_c = \arcsin\left(\frac{n'}{n}\right)$$


Das ist aber nur möglich wenn $n'<n$, also zum Beispiel, wenn man von unter dem Wasser an die Wasseroberfläche schaut, oder Licht in ein Glasfaser einkoppelt.

### 1.1.2 Fermatsche Prinzip


<b>Der tatsächliche Weg, den ein Lichtstrahl bei seiner Ausbreitung zwischen zwei gegebenen Punkten in einem optischen System nimmt, ist so beschaffen, dass die optische Weglänge bei Variation des Lichtweges ein Extremum darstellt.</b>

Wir berechnen wiederum das Snellius'sche Brechungsgesetz, diesmal aber benuten wir das Fermatsche Prinzip:


<img src="figures/1.1_7.png" alt="1.1_7.png" style="width: 400px;"/>

Die Dauer die das Licht zwischen $A$ and $C$ zurücklegt ist

$$
t=\frac{\overline{AB}}{c}+\frac{\overline{BC}}{c'}
$$

Mit Pythagoras gefüllt ergibt sich
$$
t=\frac{\sqrt{z_1^2+h^2}}{c}+\frac{\sqrt{z_2^2+(b-h)^2}}{c'}
$$

Extremum 

$$
\frac{\mathrm{d}t}{\mathrm{d}h}\overset{!}{=}0
$$

Dann
\begin{eqnarray*}
\frac{\mathrm{d}t}{\mathrm{d}h}&=&\frac{h}{c\sqrt{z_1^2+h^2}}-\frac{b-h}{c'\sqrt{z_2^2+(b-h)^2}}\\
&=&\frac{1}{c}\sin \epsilon-\frac{1}{c'}\sin \epsilon'=0
\end{eqnarray*}

und ist identisch mit dem Snellius'sche Gesetz:

$$
n\,\sin\epsilon = n'\,\sin\epsilon'\quad .
$$


Ganz Allgemein:

Bei geringer Änderung des Lichtweges, soll die optische Weglänge gleich bleiben:

$$
t=\int \frac{\mathrm{d}s}{c} = \frac{1}{c_0}\int n(s)\,\mathrm{d}s = \mathrm{Extr.}
$$


### 1.1.3 Abbildung durch ein optisches System

<img src="figures/1.1_8.png" alt="1.1_8.png" style="width: 500px;"/>

Im Objektraum divergieren die Wellenfronten, im Bildraum konvergieren die Wellenfronten.

Strahlen, die die gleiche Zeit benötigen heißen *isochron*, dies sind nach Fermat alle Strahlen, die durch ein optisches System laufen.

Umkehr der Zeit $\rightarrow$ $O'\leftrightarrow O$ sind konjugierte Punkte des optischen Systems.

<b>Bildfehler:</b>

* nicht alle Strahlen gehen von $O$ nach $O'$, dadurch wird das Bild dunkler
* Punkte nicht von $O$ gehen zu $O'$, dieser Bildfehler nennt sich Aberration
* Jedes optische System ist beugungsbegrenzt

### 1.1.4 Ideale Abbildung 

Eine fehlerfreie Abbildung, die den Objektpunkt $O$ in den Bildpunkt $O'$ überführt.

<img src="figures/1.1_9.png" alt="1.1_9.png" style="width: 400px;"/>

Die Laufzeit eines Strahls durch ein Medium mit Dicke $d$ und Brechzahl $n$ ist

$$t =\frac{d}{c}=\frac{n\,d}{c_0}$$

$n\cdot d$ wird die optische Weglänge genannt.

Alles Strahlen müssen die gleiche Laufzeit haben (Fermatsches Prinzip), also die gleich optische Weglänge, für obiges Beispiel:

$$
n\, l_1 +n'\,l_1'=n\,l_2+n'\,l_2'=\mathrm{konst.}
$$


Wir beschreiben die brechende Fläche $\Sigma$ durch die Funktion $P(y,z)$

Wiederum mit Hilfe von Herrn Pythagoras:

$$
n\left(y^2+z^2\right)^{1/2}+n'\left(y^2+\left(l_2+l_2'-z\right)^2\right)^{1/2}=\mathrm{konst}\tag{1.1}
$$

Die Grössen $l_2$ und $l_2'$ sind Konstanten des optischen Systems. Die Fläche die mit Gl.(1.1) beschrieben wird ist ein *kartesisches Ovaloid*.

Der Bildpunkt befindet sich nun aber im Gebiet mit Brechungsindex $n'$. 


Ein optisches System, dass in $n$ endet nennt sich Linse. Diese besteht aus zwei brechenden Flächen.

### 1.1.5 Konvention:
* konjugierte Punkte erhalten den gleichen Buchstaben, Bildseite mit zusätzlichem Strich, z.B. $y$ und $y'$
* Objektbrennweite und Objektbrennpunkt wird mit $\overline{f}$ und $\overline{F}$, der Bildbrennpunkt und die Bildbrennweite mit $F'$ und $f'$ (bemerke den Strich!)
* Lichtrichtung verläuft immer von links nach rechts
* $n$ ist vor der brechenden Fläche, $n'$ hinter der Fläche
* Strecken werden vom Bezugspunkt gemessen und haben eine Pfeilrichtung
* Strecken nach links und unten sind negativ.
* Winkel mit griechischen Buchstaben, z.B. $\sigma$ und $\sigma'$
* Vorzeichen der im mathematischen Sinn vom Bezugsschenkel
* Schnittwinkel ist der Bezugsschenkel die optische Achse, Einfalls/Ausfallswinkel ist das Lot.
