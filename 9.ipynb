## 9. Beugung

### 9.1 Fourier Transformierte in 2 Dimensionen

Zur Wiederholung, die Fourier Transformierte einer Funktion ist

$$A(\nu)={\cal F}[f(t)](\nu)=\int_{-\infty}^{+\infty}f(t)\,e^{-2\pi\,i\,\nu t}\mathrm{d}t\quad .$$

Die Fourier Transformierte $A(\nu)$ wird komplexe Spektralfunktion genannt.

Wir können die Fourier Tranformation auch für Funktionen auf 2 Dimensionen erweitern. Statt $t$ nehmen wir nun die Variablen $x$ und $y$ und die entsprechenden Frequenzen nennen wir $\nu_x$ und $\nu_y$. 

\begin{eqnarray*}
\overline{E}\left(\nu_x,\nu_y\right)&=&{\cal F}[E(x,y)]\left(\nu_x,\nu_y\right)\\
&=&\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}\,E(x,y) \,e^{-2\pi\,i(\nu_x \,x+\nu_y\,y)}\mathrm{d}x\,\mathrm{d}y
\end{eqnarray*}

### 9.2 Skalare Beugungstheorie

Wir geben die allgemeine Lösung der Ausbreitung einer ebenen elektrischen Welle $E_e(x,y)$ die auf eine eine komplexe Durchlässigkeit $\tau(x,y)$ am Ort $z=0$ trifft.

<img src="figures/9.2_1.png" alt="9.2_1.png" style="width: 550px;"/>

Am Ort $z=0$ bezeichnen wir das elektrische Feld 

$$E(x,y)=\tau(x,y)\,E_e(x,y)$$

Wenn wir die Ausrichtung des elektrischen Feldes (Polarisation) nicht berücksichtigen, gibt sich aus den Maxwell Gleichungen für das skalare E-Feld folgende Lösung im Abstand $z$:

$$E(x,y)=\frac{1}{i\lambda}\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}
E(x,y)\,\frac{e^{k\,r}}{r}\,\cos(\vec{\hat{n}},\vec{\hat{r}})\,\mathrm{d}x\,\mathrm{d}y\tag{9.2.1}$$

Bei genauer Betrachtung ist diese sehr Ähnlich zum Huygenschen Prinzip. An jedem Punkt $z=0$ werden Kugelwellen ausgestrahlt. Der Faktor $1/\lambda i$ ist eine Phasenverzoegerung und Normierung, und der Kosinusterm im Integral gibt eine Vorzugsrichtung in positive und negative Richtungen an, ähnlich wie ein Dipol Strahler.

Dieses allgemeine (Kirchhoffsche) Beugungsintegral kann fast nicht analytisch gelöst werden, selbst auf dem Computer benötigt man wegen der hohen Frequenz des Arguments in der Exponential-Funktion eine sehr lange Rechenzeit. Wir führen nun zwei Näherungen ein, die erste Näherung ist die Fresnel und die zweite die Fraunhofer Näherung, die beide mit Hilfe der 2-dimensionalen Fourier Transformation beschrieben werden können. Die Fraunhofer Näherung ist besonders interessant, da es hier eine sehr einfachen Zusammenhang gibt.

### 9.3 Fressnel Näherung

Die Frensel Näherung ist eine paraxiale Näherung, d.h.

$$|x|,|y|\ll z$$
$$|x'|,|y'|\ll z$$
$$\cos(\vec{\hat{n}},\vec{\hat{r}})=1$$

Weiterhin nehmen wir an, dass die Amplitude der Kugelwelle proportional mit $z$ abfällt.

Den Exponenten in Gleichung (9.2.1) müssen wir aber für kleine $x/z$ und $y/z$ mit Taylor entwickeln.

\begin{eqnarray*}
r&=&\sqrt{\left(x-x'\right)^2 + \left(y-y'\right)^2+z^2}=
z\sqrt{1+\frac{\left(x-x'\right)^2}{z^2}+\frac{\left(y-y'\right)^2}{z^2}}\\
&\approx& z\left(1+\frac{1}{2}\frac{(x-x')^2}{z^2}+\frac{1}{2}\frac{(y-y')^2}{z^2}\right)=
z+\frac{1}{2}\frac{(x-x')^2}{z}+\frac{1}{2}\frac{(y-y')^2}{z}
\end{eqnarray*}

Eingesetz in Gleichung (9.2.1) ergibt sich

$$E(x',y',z)=\frac{e^{ikz}}{i\lambda\,z}\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}
E(x,y)\,e^{{\frac{ik}{2z}}\left[\left(x-x'\right)^2+\left(y-y'\right)^2\right]}\,\mathrm{d}x\,\mathrm{d}y\tag{9.3.1}$$

Da nicht über $x'$ und $y'$ integriert wird können wir einen Teil aus dem Integral herausziehen:

$$E(x',y',z)=\frac{e^{ikz}}{i\lambda\,z}\,e^{i\frac{\pi}{\lambda z}\left(x'^2+y'^2\right)}\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}
E(x,y)\,e^{\frac{i\pi}{\lambda z}\left( x^2+y^2 \right)}
e^{-2\pi i\left(\frac{x'}{\lambda z}x+\frac{y'}{\lambda z}y
\right)}\,\mathrm{d}x\,\mathrm{d}y\tag{9.3.2}$$

Und wenn wir die Funktion $A(x',y',z)$ einführen

$$A(x',y',z)=\frac{e^{ikz}}{i\lambda\,z}\,e^{i\frac{\pi}{\lambda z}\left(x'^2+y'^2\right)}$$

kann Gleichung (9.3.2) als Fourier Transformation geschrieben werden:

$$E(x',y',z)=A(x',y',z)\,{\cal F}\left[E(x,y)\,e^{\frac{i\pi}{\lambda z}\left( x^2+y^2 \right)}\right]\left(\frac{x'}{\lambda z},\frac{y'}{\lambda z}\right)\tag{9.3.3}$$

Gleichung (9.2.4) ist die Fresnel Näherung des skalaren Beugungsintegrals. Diese sagt aus, dass in paraxialer Näherung das elektrische Feld sich wie eine Fourier Transformierte des Ausgangsfeldes - multipliziert mit einem quadratischen Phasenfaktor - verhält.
Allgemein gibt die Fresnel Näherung eine gute Lösung ab ca. 10 Wellenlängen.

#### Beispiel: Beugung an einer Kante

Das Fresnelsche Beugungsintegral eine Kante by $x=0$ ist 

\begin{equation}
E(x',z)=E_0\frac{ e^{i k z}}{i\,\lambda z}\int_{0}^{+\infty} 
e^{\frac{i k}{2z}\left(x-x'\right)^2} \,dx\quad.
\tag{6a}
\end{equation}

Das Integral lässt sich analytisch lösen und man erhält nach längerer Rechnung mit

\begin{equation}
w=\sqrt{\frac{2}{\lambda z}} x'\tag{6b}
\end{equation}

die Intensitätsverteilung

\begin{equation}
I \propto \frac{1}{2}\left(\left[\frac{1}{2}+\mathcal{C}(w)\right]^2+
\left[\frac{1}{2}+\mathcal{S}(w)\right]^2\right) \quad .\tag{6c}
\end{equation}

Die beiden Funktionen $\mathcal{C}$ and $\mathcal{S}$ sind die sogenannten [Fresnel Integrale](https://en.wikipedia.org/wiki/Fresnel_integral).

<em> Anmerkung: Details der Rechnung finden Sie im Lehrbuch.</em>

Hier sehen Sie die Intensitätsverteilung auf einem Schirm im Abstand $z$ von der Kante.

```python
%matplotlib inline
import numpy as np
from scipy.special import fresnel
from ipywidgets import interact,widgets
import matplotlib.pyplot as plt
from IPython.display import display



def pltborder(z):
    ll=532e-9
    x = np.linspace(-4,10,1000)*ll
    s,c=fresnel(x*(2/ll/(z*ll))**.5)
    y =.5*((.5+c)**2+(.5+s)**2)
    plt.plot(x/ll, y,linewidth=2);
    plt.gca().tick_params(axis='x', labelsize=20)
    plt.gca().tick_params(axis='y', labelsize=20)
    plt.xlabel(u'x\' (lambda)',fontsize=20)
    plt.ylabel(u'Intensität (a.u.)',fontsize=20,)
    
interact(pltborder, z=widgets.FloatSlider(min=1,max=100,step=1,value=50,description=u'Abstand z (λ)'));
```

### 9.4 Fraunhofer Näherung

Wenn der Abstand $z$ noch wesentlich größer als 10 ist (das Fernfeld), dann kann der quadratische Phasenfaktor in Gl. (9.3.3) vereinfacht werden. D.h. wenn 

$$z\gg \frac{\pi}{\lambda}\left(x^2+y^2\right) \quad \rightarrow \quad e^{\frac{i\pi}{\lambda z}\left( x^2+y^2 \right)}\approx 1$$

Dann ist 

$$E(x',y',z)=A(x',y',z)\,\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}
E(x,y)e^{-2\pi i\left(\frac{x'}{\lambda z}x+\frac{y'}{\lambda z}y
\right)}\,\mathrm{d}x\,\mathrm{d}y\tag{9.4.1}$$

Und Gleichung (9.4.1) ist eine Fourier Transformation in den Variablen $\nu_x=\frac{x'}{\lambda z}$ und  $\nu_y=\frac{y'}{\lambda z}$:

$$E(x',y',z)=A(\lambda z \nu_x,\lambda z \nu_y,z)\,{\cal F}\left[E(x,y)\right]\left(\nu_x,\nu_y\right)\tag{9.4.2}\quad .$$

Diese Ergebnis zeigt, dass im Fernfeld, z.B. bei der Abbildung mit einer positiven Linse ins Unendliche, wir das resultierende Beugungsfeld eine Fourier Transformierte ist (bis auf einen Amplitudenfaktor).

Die Bedeutung der Raumfrequenzen $\nu_x$ und $\nu_y$ werden aus unten stehender Abbildung ersichtlich.

<img src="figures/9.4_1.png" alt="9.4_1.png" style="width: 550px;"/>

$$\nu_x=\frac{x'}{\lambda\, z}=\frac{\tan \alpha}{\lambda}\approx \frac{\alpha}{\lambda}$$

und 


$$\nu_y=\frac{y'}{\lambda\, z}=\frac{\tan \beta}{\lambda}\approx \frac{\beta}{\lambda}$$


Wir betrachten uns jetzt genauer den Exponentialterm in Gl. (9.4.1) und schreiben 

$$\frac{x'}{z}=\tan{\alpha}\approx\sin \alpha$$
bzw.
$$\frac{y'}{z}=\tan{\beta}\approx\sin \beta$$

dann sehen wir, dass der Term ein ebene Welle in $z$-Richtung und den Winkeln $\alpha$ zur $x$-Achse und $\beta$ zur $y$-Achse ausbreitet:

$$e^{i\,k\,z}\,e^{-2\pi i\left(\frac{x'}{\lambda z}x+\frac{y'}{\lambda z}y
\right)}=e^{-i\,k\left(\sin\alpha x + \sin\beta y - z\right)}\quad .$$

Kleine Winkel entsprechen kleinen Raumfrequenzen und große Winkel hohe Raumfrequenzen.

Wir berechnen nun das Intensitätsmuster durch die Multiplikation des komplex konjugierten elektrischen Feldes. Es gilt allgemein, dass

$$e^{i\varphi}\,e^{-i\varphi}=1$$

daher wird aus der komplexe Amplitudenfunktion $A\,A^*=(\lambda\,z)^{-2}$ und wir erhalten fuer die Intensität


$$I(\nu_x,\nu_y)=\frac{1}{\lambda^2\,z^2}\,\left|{\cal F}\left[E(x,y)\right]\left(\nu_x,\nu_y\right)\right|^2\tag{9.4.3}\quad .$$


### 9.5 Fourier Transformation mit einer positiven Linse

Eine positive Linse fokussiert ein paralleles Lichtbündel auf einen Punkt in der hinteren Fokusebene. Die Koordinate in der hinteren Brennebene ist $(u,v)$. Die Abbildung unten zeigt die Koordinate $v$:

<img src="figures/9.5_1.png" alt="9.5_1.png" style="width: 550px;"/>


Die Abbildunggleichung ergibt

$$\frac{v}{f_l}=\tan\beta\approx\beta$$

und analog für $u$ und $\alpha$:

$$\frac{u}{f_l}=\tan\alpha\approx\alpha$$


Das bedeutet, dass eine Linse das Fraunhofer Beugungsbild in der hinteren Brennebene erzeugt, da

$$\nu_x=\frac{x'}{\lambda \,z}\approx\frac{\alpha}{\lambda}=\frac{u}{\lambda\,f_l}$$

$$\nu_y=\frac{y'}{\lambda \,z}\approx\frac{\beta}{\lambda}=\frac{v}{\lambda\,f_l}$$


Wir können nun das Fraunhofer Beugungsintegral in als Funktion von $u=\lambda f_l \nu_x$ und $v=\lambda f_l \nu_y$ schreiben:

$$\overline{E}(u,v)=A(u,v,f_l)\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}E(x,y)\,e^{-2\pi\,i\left(\frac{u}{\lambda\,f_l}x+\frac{v}{\lambda\,f_l}y\right)}\,\mathrm{d}x\,\mathrm{d}y\quad .$$


Wiederum fällt der Phasenfaktor $A$ bei der Intensitätsbildung heraus und wir erhalten

$$I(u,v)=\left|\overline{E}(u,v)\right|^2\propto \left|{\cal F}\left[ E\left(x,y\right)\right]\right|^2\quad . $$

Ein Linse führt eine optische Fouriertransformation durch. Das Beungungsbild der Eingangsfeldverteilung in der hinteren Brennebene wird einfach "Spektrum" genannt. Große Brennweiten helfen das Spektrum zu vergrößern oder leichter sichtbar zu machen.

### 9.6 Beispiele der optischen Fourier Transformation

#### 9.6.1 Punktlichtquelle

$$E(x,y)=E_0\,\delta(x-x_0)\delta(y-y_0)$$

Die Fourier Transformation ergibt

\begin{eqnarray*}
{\cal F}[E(x,y)]&=&\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}E_0\,\delta (x-x_0)\delta (y-y_0)\,e^{-2\pi\,i\left(\nu_x x+ \nu_y y \right)}\,\mathrm{d}x\,\mathrm{d}y\\
&=&E_0\,e^{-2\pi\,i\left(nu_x\,x_0+\nu_y\,y_0\right)}
\end{eqnarray*}

Dies ist ein gleichmäßig verteile ebene Welle. Wenn sich die punktförmige Lichtquelle am Ursprung $(x_0,y_0)=(0,0)$ befindet erhalten wir eine ebene Welle entlang der $z$-Achse, sonst unter einem Winkel. Die Helligkeitsverteilung oder das Spektrum in beiden Fällen ist identisch, nämlich $|{\cal F}[E(x,y)]|^2=|E_0|^2$.

<img src="figures/9.6_1.png" alt="9.6_1.png" style="width: 550px;"/>

#### 9.6.2 Ebene Welle

Eine Ebene Welle kann  mit einer konstanter Amplitude $E_0$ beschrieben werden:

$$E(x,y)=E_0$$

Dann ergibt die Fourier Transformation:

\begin{eqnarray*}
{\cal F}[E(x,y)]&=&\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}E_0\,\,e^{-2\pi\,i\left(\nu_x x+ \nu_y y \right)}\,\mathrm{d}x\,\mathrm{d}y\\
&=&E_0\,\delta(\nu_x)\,\delta(\nu_y)
\end{eqnarray*}


Bildlich dargestellt ist dies:

<img src="figures/9.6_2.png" alt="9.6_2.png" style="width: 370px;"/>

#### 9.6.3 Unendlich langer dünner Spalt

Ein unendlich langer Spalt in vertikaler ($y$) Richtung ist durch solch eine Feldverteilung gegeben:

$$E(x,y)=E_0\delta(x)\quad .$$

\begin{eqnarray*}
{\cal F}[E(x,y)]&=&E_0\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}\delta(x)\,\,e^{-2\pi\,i\left(\nu_x x+ \nu_y y \right)}\,\mathrm{d}x\,\mathrm{d}y\\
&=&E_0\int_{-\infty}^{+\infty}\,e^{-2\pi\,i\nu_y y}\mathrm{d}y=E_0\,\delta(\nu_y)\quad .
\end{eqnarray*}


Das Spektrum ist um $90^\circ$ gedreht.

<img src="figures/9.6_3.png" alt="9.6_3.png" style="width: 500px;"/>


#### 9.6.4 Unendlich langer breiter Spalt

Die Feldverteilung eines Spaltes kann geschrieben werden als


$$E(x,y)=E_0\, \mathrm{rect}\left(\frac{x}{a}\right)=E_0\left\{
\begin{array}{l}
1\quad |x|<a/2\\
0\quad \mathrm{sonst}\\
\end{array}
\right. \quad .$$

\begin{eqnarray*}
{\cal F}[E(x,y)]&=&E_0\int_{-\infty}^{+\infty}\int_{-a/2}^{+a/2} e^{-2\pi\,i\left(\nu_x x+ \nu_y y \right)}\,\mathrm{d}x\,\mathrm{d}y\\
&=&E_0\int_{-\infty}^{+\infty}e^{-2\pi\,i\,\nu_y y}\,\mathrm{d}y\,\int_{-a/2}^{+a/2}e^{-2\pi\,i\,\nu_x x}\,\mathrm{d}x\\
&=&E_0\delta(\nu_y)\frac{1}{-2\pi i \nu_x}\left[e^{-2\pi i \nu_x a/2}-e^{2\pi i \nu_x a/2}\right]\\
&=&E_0\delta(\nu_y)\frac{\sin(\pi \nu_x a)}{\pi \nu_x}=E_0\,a_0\delta(\nu_y)\,\mathrm{sinc}(\pi \nu_x a)\quad .
\end{eqnarray*}

Die sinc-Funktion ist folgendermaßen definiert:

$$\mathrm{sinc}(x)=\frac{\sin\pi x}{\pi x}\quad .$$

<img src="figures/9.6_4.png" alt="9.6_4.png" style="width: 550px;"/>

```python
%matplotlib inline
import numpy as np
from scipy.special import fresnel
from ipywidgets import interact,widgets
import matplotlib.pyplot as plt
from IPython.display import display

ll=532e-9
z=100.*ll
x = np.linspace(-50,50,1000)*ll
a=6.*ll
y=(np.sin(np.pi*x*a/z/ll)/(np.pi*x*a/z/ll))**2.
y=y/y.max()#x/ll/z*a*ll
plt.plot(x/ll/z*a, y,linewidth=2);
ax1=plt.gca()
ax1.tick_params(axis='x', labelsize=20)
ax1.tick_params(axis='y', labelsize=20)
plt.xlabel(u'x\' (lambda*z/a)',fontsize=20)
plt.ylabel(u'Intensität (a.u.)',fontsize=20);
```

#### 9.6.5 Zusammengesetzte beugende Strukturen

Linearität der Fourier Transformation:

$${\cal F}[E_1(x,y)+E_2(x,y)](\nu_x,\nu_y)={\cal F}[E_1(x,y)](\nu_x,\nu_y)+{\cal F}[E_2(x,y)](\nu_x,\nu_y)\quad .$$

Verschiebungseigenschaft der Fourier Transformation:

$${\cal F}[E(x+\Delta x,y+\Delta y)](\nu_x,\nu_y)=e^{2\pi\,i\left(\nu_x \Delta x+\nu_y \Delta y\right)}\,{\cal F}[E(x,y)](\nu_x,\nu_y)$$

Faltungseigenschaft der Fourier Transformation:

$${\cal F}[E_1(x,y) \cdot E_2(x,y)](\nu_x,\nu_y)={\cal F}[E_1(x,y)]*{\cal F}[E_2(x,y)]\quad .$$

und umgekehrt, wenn die Feldverteilung eine Faltung ist

$$E(x,y)=E_1(x,y) * E_2(x,y)=\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}E_1(\xi,\eta)\,E_2(x-\xi,y-\eta)\,\mathrm{d}\xi\,\mathrm{d}\eta\quad ,$$

dann ist die Fourier Transformierte das Produkt der individuellen Fourier Transformierten.

$${\cal F}[E_1 * E_2](\nu_x,\nu_y)={\cal F}[E_1(x,y)](\nu_x,\nu_y)\cdot{\cal F}[E_2(x,y)](\nu_x,\nu_y)\quad .$$

#### 9.6.6 Senkrechtes Gitter
Das senkrechte Gitter besteht aus $M$ Spalte mit Breite $a$ und Gitterabstand $d$ ($d>a$) in der $x$-Richtung. Die Feldverteilung hinter dem Gitter kann so beschrieben werden:

$$E(x,y)=E_0\sum_{m=1}^{M}\mathrm{rect}\left(\frac{x}{a}-\frac{m d}{a}\right)=E_0\left[\sum_{m=1}^{M}\delta(x-md)\right]*\mathrm{rect}\left(\frac{x}{a}\right)\quad .$$

Wir können jetzt den Linearitäts-, Faltungs- und Verschiebungsatz der Fouriertransformation anwenden und erhalten:

\begin{eqnarray*}
{\cal F}[E]&=&E_0\,\delta(\nu_y)\frac{\sin(\pi a \nu_x)}{\pi \nu_x}\sum_{m=1}^{M}e^{-2\pi i\, m\, d\, nu_x}\\
&=&E_0\delta(\nu_y) a \,\mathrm{sinc}(a\nu_x)\,e^{-\pi i\,d\,\nu_x\left(M+1\right)}
\frac{\sin(\pi M d \nu_x)}{\sin ( \pi d \nu_x)}
\end{eqnarray*}

<img src="figures/9.6_5.png" alt="9.6_5.png" style="width: 500px;"/>


Bei der Berechnung der Intensität fällt der Phasenfaktor raus:

$$I(\nu_x,\nu_y)=|E_0|^2\delta(\nu_y) a^2 \mathrm{sinc}^2(a\,\nu_x)\frac{\sin^2(\pi M d \nu_x)}{\sin^2( \pi d \nu_x)}$$

Die grobe Strutur der Intensitätsverteilung ist proportional zu $a^2 \mathrm{sinc}^2(a\,\nu_x)$ also eine Funktion des individuellen Spaltes. Die periodische  Wiederholung des Spektrum kommt von dem Faktor $\sin^2(\pi M d \nu_x)/\sin^2( \pi d \nu_x)$, d.h. diese Feinstruktur hängt von der Größe des Gitters ab.
