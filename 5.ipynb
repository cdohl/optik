## 5. Optische Instrumente

### 5.1 Kamera
Das einfachste optische Instrument ist die Lochkamera. Die Lichtstrahle treten durch ein kleines Loch in ein lichtdichtes Gehäuse und treffen entweder auf einen Film oder eine Mattscheibe. Die Lochkamera fokussiert nicht sondern erlaubt, dass nur "wenige" Strahlen vom Objekt auf die Mattscheibe treffen. 
<img src="figures/5.1_1.png" alt="5.1_1.png" style="width: 600px;"/>
Das Loch sollte idealerweise klein und kreisförmig sein. Wird das Loch durch eine positive Linse ersetzt, erhält man eine gewöhnliche Kamera. Der Vorteil ist die erhöhte Bildhelligkeit und Bildschärfe. Der Abstand zur Bildweite ist nun durch die Abbildungsgleichung von der Objektweite und der Brennweite der Linse abhängig. Das Scharfstellen geschieht durch Änderung des Abstandes der Linse von der Bildebene. Ein *Weitwinkelobjektiv* wird eine Linse/ein Linsensystem mit kurzer Brennweite und große Sehwinkel bezeichnet. Ein *Teleobjektiv* ist ein Objektiv mit großer Brennweite und hoher Vergrößerung.


#### 5.1.1 Blende

Die Blende bestimmt die Menge and durchgelassenem Licht und damit die Belichtungszeit. Die Blende beschränkt den nutzbaren Durchmesser der Linse, s. Abb. unten:

<img src="figures/5.1_2.png" alt="5.1_2.png" style="width: 500px;"/>

Die Bestrahlungstärke 
$$E_e \propto \frac{\mathrm{Fläche\,des\,durchgelassenen\,Lichtes}}{\mathrm{Fläche\,des\,Bildes}}=\frac{D^2}{d^2}$$

Die Bildgröße $d$ ist proportional zur bildseitigen Brennweite $f'$

$$E_e \propto \frac{D^2}{f'^2}$$ 

und man definiert die Blendenzahl 

$$k\equiv\frac{f'}{D}\quad,$$
wobei $D$ der Durchmesser der Eintrittspupille der Linse ist. Auf dem Objektiv wird nicht die Blendenzahl sondern die relative Öffnug angegeben, also statt $k=5.6$ das Verhältnis $1:5.6$. Die Bstrahlungsstärke is demnach

$$E_e \propto \frac{1}{k^2}\quad .$$ 

| Blendenzahl $k$ | $k^2$ | relative Bestrahlungsstärke $E_e$
|----|-----|-----|
| 1  | 1   |1    |
|1.4| 2    |1/2  | 
|2| 4    |1/4  | 
|2.8| 8    |1/8  | 
|4| 16    |1/16  | 
|5.6| 32    |1/32  | 
|8| 64   |1/64 |
|11|128  |1/128 |
|16|256    |1/256  | 

Wird die Blendenzahl $k$ um eine Stufe erhöht, verdoppelt sich die Belichtungszeit. 

#### 5.1.2 Schärfentiefe

<img src="figures/5.1_3.png" alt="5.1_3.png" style="width: 500px;"/>

Dear Bildpunkt in $a_0$ wird als Bildpunky $a'_0$ scharf abgebildet. Bildpunkte vor und hinter $a_0$ werden  in der Bildebene $a'_0$ unscharf als Kreisscheiben abgebildet mit Durchmesser $u'$ abgebildet. 

Nach kurzer Ableitung (s. Kapitel 6.3) im Lehrbuch erhalten wir, dass ein Punkt im Abstand $a_v$ und $a_h$ zu einer Kreisschebe mit Durchmesser von $u'$ führt.

$$a_v=\frac{a_0\,f'^2}{f'^2-u'\,k\left(a_0+f'\right)}\tag{5.1.1}$$
und
$$a_h=\frac{a_0\,f'^2}{f'^2+u'\,k\left(a_0+f'\right)}\tag{5.1.2}$$

Gleichungen (5.1.1) und (5.1.2) lassen sich vereinfachen, wenn wir annehmen, dass die bildseitige Brennweite $f'\ll |a_0|$ und $u'\ll f'$:

$$\frac{1}{a_v}=\frac{1}{a_0}-\frac{u'\,k}{f'^2}\tag{5.1.3}$$

und 

$$\frac{1}{a_h}=\frac{1}{a_0}+\frac{u'\,k}{f'^2}\tag{5.1.4}\quad .$$

Beide Gleichungen lassen sich addieren und wir erhalten den einfachen Ausdruck 

$$\frac{1}{a_v}+\frac{1}{a_h}=\frac{2}{a_0}$$

Daraus folgt, dass wenn die hintere Schärfentiefe $a_h \rightarrow -\infty$ geht, die vordere Schärfentiefe $a_v=a_0/2$. 

Mit dem kleinen Programm können Sie die Schäfentiefe einer Kamera bestimmen und sie sehen die Abhängigkeiten von der Bildweite, Blendenzahl, Brennweite und dem Unschärfekreis. 

```python
from ipywidgets import *
import numpy as np

a0=-5.   #Objectweite in m
        #Brennweite fp in mm

def update(a0, k, fp, up):
    fpm=fp*1e-3 #Brennweite fp in mm
    upm=up*1e-6 #Unschärfekreis in um
    
    av=a0*(fpm)**2./(fpm**2.-upm*k*(a0+fpm))
    ah=a0*(fpm)**2./(fpm**2.+upm*k*(a0+fpm))
    
    print("ah: {0:10.4f} m < a0 < av: {1:10.4f} m".format(ah,av))
    

interact(update,a0=widgets.FloatSlider(min=-100., max=-1., step=1., value=-5., description='a0 [m]'),
         k=widgets.FloatSlider(min=1.,max=16.,value=5.6,description='Blendenz. k'),
         fp=widgets.FloatSlider(min=20.,max=200.,value=50., description='f\' [mm]'),
         up=widgets.FloatSlider(min=1.,max=100,value=30., description='Unsch. [um]'));
```

### 5.2 Lupe

Das Auge kann bis zu einem Abstand von 25cm akkomodieren. Ein kleines Objekt wie in der Abbildung unter mit Höhe $y$ erscheint unter dem Sehwinkel $w_s$.

<img src="figures/5.2_1.png" alt="5.2_1.png" style="width: 500px;"/>

Die Lupe vergrößert diesen Sehwinkel, und damit auch die Größe des Objektes auf der Netzhaut. Die Winkelvergrößerung 

$$\Gamma'=\frac{\tan w'_s}{\tan w_s}\tag{5.2.1}$$

wobei $w_s$ der Sehwinkel ohne und $w'_s$ der Sehwinkel mit Lupe ist. Dazu muss ein virtuelles Bild erzeugt werden, was durch die Positionierung des Objektes zwischen der objektseitigen Brennweite und der Lupe im Abstand $a$ erreicht wird. Zur Konstruktion der Strahlen benötigen Sie den Brennpunkt, der nicht eingezeichnet ist.  

<img src="figures/5.2_2.png" alt="5.2_2.png" style="width: 500px;"/>

Der Winkel $\tan w_s=y/(-a_s)$ und der Winkel $\tan w'_s=y'/(-a_s+e_A)$ ergeben sich aus obiger Abbildung. Ferner mit dem Strahlenstatz erhalten wir $\frac{y}{a}=\frac{y'}{a'_L}$. Eingesetzt in Gl. (5.2.1) erhalten wir als Winkelvergrößerung:

$$\Gamma'=\frac{y'}{y}\frac{-a_s}{e_A-a'_L}=-\frac{f'-a'_L}{e_A-a'_L}\frac{a_s}{f'}\tag{5.2.2}\quad .$$

Lassen wir nun $a'_L$ groß werden, das heisst das Auge schaut sich das Objekt unakkommodiert an dann vereinfacht sich Gleichung (5.2.2) zur Lupenvergrößerung

$$\Gamma'_L=-\frac{a_s}{f'}=\frac{0.25\,\mathrm{m}}{f'}\tag{5.2.3}\quad .$$

Wir das Auge auf den Nahpunkt (Bezugssehweite) maximal akkommodiert ist $a_s=a'_L-e_A$ (Vorzeichen!) ergibt sich eine stärkere Vergrößerung.

$$\Gamma'=1-\frac{a_s}{f'}-\frac{e_A}{f'}=1+\Gamma'_L-\frac{e_A}{f'}\tag{5.2.4}\quad .$$


<b> Beispiel Lupe </b>

Brennweite der Linse ist $f'=20\,$cm. Wie groß ist die Lupenvergrößerung bei Akkommodation auf die Bezugsfernweite und Augenabstand $e_A=10\,$cm und $e_A=0\,$cm?

Die Lupenvergrößerung ist 

$$\Gamma'=\frac{25\,\mathrm{cm}}{f'}=1.25$$

Die beiden Vergrößerungen können wir mit Gleichung (5.2.4) berechnen.

$$\Gamma=1+1.25-\frac{10\,\mathrm{cm}}{20\,\mathrm{cm}}=1.75$$
und
$$\Gamma=1+1.25=2.25$$


### 5.3 Mikroskop

#### 5.3.1 Aufbau
Das Mikroskop besteht aus mindestens zwei positive Linsen, einer Objektivlinse und einer Lupe oder Okular. Die Objektivlinse hat eine kleine Brennweite. Das Objektiv erzeugt ein reelles Zwischenbild. Das Okular erzeugt nun ein virtuelles Bild entweder im Unendlichen oder innerhalb der Akkommodationsweite des Auges. Dafür liegt die objektseitige Brennweite des Okulars entweder auf dem Zwischenbild oder kurz davor. Weiterhin hat man sich auf eine festen Abstand zwischen Objektiv und Okular (die Tubuslänge $t_{me}) von $16\,$cm geeinigt. 
Der Strahlengang ist in der unten stehenden Abbildung gezeichnet.

<img src="figures/5.3_1.png" alt="5.3_1.png" style="width: 650px;"/>

Für ein Bild im Unendlichen ist die Winkelvergrößerung der einer Lupe, Gl. (5.2.3), mit

$$\Gamma'_M=-\frac{a_s}{f'_t}=\frac{25\,\mathrm{cm}}{f'_t}\tag{5.3.1}$$

$f'_t$ ist die Brennweite der zwei Linsen. Wir habe diese in Kapitel 2.5, Gl. (2.5.1), berechnet.

$$\frac{1}{f'_t}=\frac{1}{f'_{Ob}}+\frac{1}{f'_{Ok}}-\frac{t_{me}}{f'_{Ok}\,f'_{Ob}}\tag{5.3.2}$$

Setzt man Gl. (5.3.2) in Gl. (5.3.1) ein

$$\Gamma'_M=\frac{-a_s\left(f'_{Ob}+f'_{Ok}-t_{me}\right)}{f'_{Ob}\,f'_{Ok}}\tag{5.3.3}$$

Diesen Ausdruck können wir vereinfachen, wenn wir Berücksichtigen, dass die Bildweite $a'_{Ob}$ des Zwischenbildes

$$a'_{Ob}=t_{me}-f'_{Ok}$$

ist. Die Bildweite $a'_{Ob}$ ist durch

$$\frac{1}{a'_{Ob}}-\frac{1}{a_{Ob}}=\frac{1}{f'_{Ob}}$$

und damit

$$\frac{a'_{Ob}}{a_{Ob}}=1-\frac{a'_{Ob}}{f'_{Ob}}=\frac{f'_{Ob}-t_{me}+f'_{Ok}}{f'_{Ob}}\tag{5.3.4}$$

Der Verhältnis von Bildweite zu Gegenstandsweite $a'_{Ob}/a_{Ob}$ wird in Gl. (5.3.3) eingesetzt:

$$\Gamma'_M=-\frac{a'_{Ob}}{a_{Ob}}\,\frac{-a_s}{f'_{Ok}}=\beta'_{Ob}\,\Gamma'_{Ok}\tag{5.3.5}$$

Die Gesamtvergrößerung ist das Produkt aus der Objektivevergrößerung und der Okularvergrößerung.

<b> Beispiel Mikroskop </b>

Berechnen Sie die Vergrößerung eines Mikroskopes mit einer Objektivbrennweite $f'_{Ob}=1.9\,$cm, einer Okularbrennweite $f'_{Ok}=5\,$cm und einer Tubuslänge von $t_{me}16.4\,$cm.

$$\Gamma'_M=-\frac{-25\,\mathrm{cm}\left(1.9\,\mathrm{cm}+5\,\mathrm{cm}-16.4\,\mathrm{cm}\right)}{1.9\,\mathrm{cm}\,\cdot\,5\,\mathrm{cm}}=-25$$

#### 5.3.3 Köhlersche Beleuchtung


Die Köhlersche Beleuchtung im Mikroskop erlaubt eine gleichmässige Ausleuchtung des Objektes und die Beleuchtung auf ein Gebiet im Sichtfeld des Objektes zu begrenzen. Dies wird durche zwei weitere Linsen, der Kollektorlinse und dem Kondensor erreicht. Zwei geschickt angebrachte Blenden erlauben das Beleuchtungsfeld und die Lichtstärke unabhängig von einander einzustellen. Zum Verständins ist es hilfreich die konjugierten Ebenen des Lukenstrahls und des Pupillenstrahls getrennt von einander anzuschauen.

<b>Lukenstrahl</b>

Das ist der Abbildung der Leuchtfeldblende (Luke). Wir sehen, dass vier konjugierte Ebenen im Mikroskop existieren. Insbesondere wird die Leuchtfeldblende auf die Objektebene abgebildet, d.h. die Blende bestimmt den beleuchteten Bereich des Objektes, die Helligkeit wird nicht beeinflusst.

<img src="figures/koehlerluke.gif" alt="koehlerluke.gif" style="width: 550px;"/>


<b>Pupillenstrahl</b>

Der Pupillenstrahl beschreibt die Abbildung der Aperturblende. Diese Ebene ist konjungiert mit der Lichtquelle, d.h. hier gibt es ein reelles Zwischenbild der Lichtquelle. Es gibt wiederum 4 konjungierte Ebenen mit der Lichtquelle. Wichtig ist, dass die Lichtquelle *nicht* auf die Netzhaut abgebildet wird. Wird nun die Aperturblende verändert, wird die Helligkeit in der Objektebene verändert.

### 5.4 Fernrohr

Es gibt Linsen und Spiegel basierte Fernrohre. Hier betrachten wir die ie Abbildung eines astronomischen oder Kepler-Fernrohres. Nahezu parallele Lichtstrahlen von weit entfernten Objekten werden durch die Objektivlinse gesammelt. Es entsteht ein reelles Bild in der Bildbrennebene $f'_{Ob}$. Die Objektivlinse ist im Durchmesser größer als die Augenpupille und damit lichtstärker. Das reelle Zwischenbild wird über das Okular wie bei einer Lube mit dem Auge beobachtet. 
<img src="figures/5.4_1.png" alt="5.4_1.png" style="width: 600px;"/>

Die Winkelvergrößerung ist

$$\Gamma'_{F\infty}=\frac{\tan w'_s}{\tan w_s}=-\frac{f'_{Ob}}{f'_{Ok}}\tag{5.4.1}\quad $$

wobei wir hier für die letzten Term den Tangens über die Größe des rellen Zwischenbildes bestimmt haben.
