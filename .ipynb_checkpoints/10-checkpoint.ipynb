## 10. Polarisation

### 10.1 Jones Vektoren

Eine elektrische Welle, die sich in $z$-Richtung ausbreitet, hat zwei transversale Komponenten, $E_x$ und $E_y$:

$$ \vec{E}=E_x\,\vec{e}_x+=E_y\,\vec{e}_y$$

Die komplexen elektrischen Feldvektoren sind:
\begin{eqnarray*}
E_x&=&\hat{E}_x\, e^{i(\omega t - kz+\varphi_x)}\\
E_y&=&\hat{E}_y\, e^{i(\omega t - kz+\varphi_y)}
\end{eqnarray*}

Die beiden Komponenten haben den Nullphasenwinkel $\varphi_x$ und $\varphi_y$.
<img src="figures/10.1_1.png" alt="10.1_1.png" style="width: 300px;"/>

\begin{eqnarray*}
\vec{E}&=&\hat{E}_x\, e^{i(\omega t - kz+\varphi_x)}\vec{e}_x+
\hat{E}_y\, e^{i(\omega t - kz+\varphi_y)}\vec{e}_y\\
&=&\left( \hat{E}_x\,e^{i\varphi_x}\vec{e}_x+\hat{E}_y\,e^{i\varphi_y}\vec{e}_y
\right)e^{i(\omega t - kz)}=\hat{\vec{E}}\,e^{i(\omega t - kz)}
\end{eqnarray*}

Der Polarisationszustand der Welle wird durch die komplexe Amplitude $\hat{\vec{E}}$ bestimmt:

$$\hat{\vec{E}}=\left(
\begin{array}{c}
\hat{E_x}\\
\hat{E_y}
\end{array}
\right)=
\left(
\begin{array}{c}
\hat{E}_x\,e^{i\varphi_x}\\
\hat{E}_y\,e^{i\varphi_y}
\end{array}\right)$$


$$\hat{E}_0=\sqrt{\hat{E}_x^2+\hat{E}_y^2}$$

Der normierte Jones-Vektor ist


$$\vec{J}=\left(
\begin{array}{c}
J_x\\ J_y 
\end{array}\right)=\frac{1}{E_0}
\left(
\begin{array}{c}
\hat{E}_x\,e^{i\varphi_x}\\
\hat{E}_y\,e^{i\varphi_y}
\end{array}
\right)\tag{10.1}$$


Der Jones Vektor hat Länge 1.



<img src="figures/10.1_2.png" alt="10.1_2.png" style="width: 500px;"/>

a) vertikal polarisiertes Licht, b) horizontal polarisiertes Licht, c) unter dem Winkel $\vartheta$ polarisiert. 

Vertikal polarisiertes Licht hat nur eine Feldkomponente in $y$-Richtung, $\hat{E}_x=0$ und wir können die Phase $\varphi_y$=0 setzen. Aus Gl. (10.1) folgt für vertikal linear polarisiertes Licht:

$$\vec{J}_v=\left(\begin{array}{c}0\\1\end{array}\right)$$

und ebenso für horizontal linear polarisiertes Lichth

$$\vec{J}_h=\left(\begin{array}{c}1\\0\end{array}\right)$$


Dass die Schwingungsebene konstant bleibt müssen die $x$ und $y$-Komponente des elektrischen Feldes die gleiche Phase $\varphi_x=\varphi_y$ haben. Wir könnnen die Bezugsphase beliebig legen und daher ist $\varphi_x=\varphi_y=0$ und erhalten als Jones Vektor: 

$$J_\vartheta=\left( \begin{array}{c}
\cos \vartheta\\
\sin \vartheta
\end{array}\right)\quad .$$

### 10.2 Zirkular polarisiertes Licht

Wenn $\hat{E}_x \ne \hat{E}_y$ und die Phasendifferenz $\delta=\varphi_y-\varphi_x$ nicht $\delta \ne 0^\circ$ oder  $\delta \ne 180^\circ$  ist, dann bekommen wir als Lösung Lissajous-Figuren. 

<img src="figures/10.1_3.png" alt="10.1_3.png" style="width: 600px;"/>


Als Beispiel nehmen wir an, dass $\hat{E_x}=\hat{E_y}=\hat{E}$ ist und dass $E_y$ $90^\circ$ gegenüber $E_x$ in der Phase nacheilt. Mit $\varphi_x=0$ ist $\delta=\varphi_y-\varphi_x=\varphi_y=-\pi/2$. Wenn $E_x$ maximale Auslenkung hat ist $E_y=0$. Eine viertel Periode ist $E_x=0$ und $E_y=\hat{E}$. Es ist also eine zirkular polarisierte Welle mit linkem Umlaufsinn.

\begin{eqnarray*}
E_x&=&\hat{E}_x\,e^{i\omega t}\\
E_y&=&\hat{E}_y\,e^{i\omega t-\pi/2}
\end{eqnarray*}

Jetzt nehmen wir den Realteil:
\begin{eqnarray*}
Re\{E_x\}&=&\hat{E}\,\cos\omega t \\
Re\{E_y\}&=&\hat{E}\,\cos\left(\omega t - \frac{\pi}{2}\right)=\hat{E}\,\sin\omega t
\end{eqnarray*}

Damit ist der Jones Vektor für linkszirkular polarisiertes Licht ist

$$\vec{J}_{lz}=
\frac{1}{E_0}
\left(
\begin{array}{c}
\hat{E}_x\,e^{i\varphi_x}\\
\hat{E}_y\,e^{i\varphi_y}
\end{array}
\right)=\frac{1}{E_0}
\left(
\begin{array}{c}
\hat{E}\,e^{i\,0}\\
\hat{E}\,e^{-i\,\pi/2}
\end{array}
\right)=
\frac{1}{\sqrt{2}}
\left(
\begin{array}{c}
1\\
-i
\end{array}
\right)
$$


Und der entsprechende Jones Vektor für rechtszirkularisierte Licht ist 

$$\vec{J}_{rz}=\frac{1}{\sqrt{2}}
\left(
\begin{array}{c}
1\\
i
\end{array}
\right)\quad .
$$


### 10.3 Elliptisch polarisiertes Licht


Jetzt nehmen wir an, dass die Phasendifferenz weiterhin $90^\circ$ beträgt, aber die Amplituden $\hat{E}_x\ne\hat{E}_y$. Der Betrag ist $\hat{E}=\sqrt{\hat{E}_x^2+\hat{E}_y^2}$. Die resultierenden Jones Vektoren sind für linkselliptisch 

$$J_{le}=\frac{1}{\hat{E}}
\left(
\begin{array}{c}
\hat{E}_x\\
-i\,\hat{E}_y
\end{array}
\right)
$$

und rechtselliptisch polarisierted Licht 

$$J_{re}=\frac{1}{\hat{E}}
\left(
\begin{array}{c}
\hat{E}_x\\
i\,\hat{E}_y
\end{array}
\right)\quad .
$$


Man kann auch elliptisch polarisiertes Licht darstellen, dass entlang der $x$- oder $y$ geneigt ist.

<img src="figures/10.1_4.png" alt="10.1_4.png" style="width: 400px;"/>

Die Polarisationsellipse ist unter dem Winkel $\alpha$ zur $x$-Achse geneigt; $\epsilon$ ist der Elliptizitätswinkel. Diesen allgemeinen Fall der Polarisation erhält man wenn
$\varphi_x=0$, $\delta=\varphi_y-\varphi_x=\varphi_y$ und $\hat{E}_x\ne\hat{E}_y$.


Dann ist der elektrische Feldvektor



$$\hat{\vec{E}}=
\left(
\begin{array}{c}
\hat{E}_x\,e^{i \varphi_x}\\
\hat{E}_y\,e^{i \varphi_y}
\end{array}
\right)=
\left(
\begin{array}{c}
\hat{E}_x\\
\hat{E}_y\,e^{i \delta}
\end{array}
\right)
$$


Wir können die $y$-Komponente in einen Realteil und einen Imaginärteil aufspalten:

$$\hat{E}_y\,e^{i\delta}=\hat{E}_y\left(\cos\delta+i\,\sin\delta\right)=\hat{E}_{yr}+i\,\hat{E}_{xr}$$


Dann ist in diesem allgemeinen Fall 


$$\hat{\vec{E}}=\left(
\begin{array}{c}
\hat{E}_x\\
\hat{E}_{yr}+i\,\hat{E}_{xr}
\end{array}
\right)
$$

mit $\hat{E}=\sqrt{\hat{E}_x^2+\hat{E}_y^2}$ erhalten wir als Jones Vektor 

$$
J=\frac{1}{\hat{E}}
\left(
\begin{array}{c}
\hat{E}_x\\
\hat{E}_{yr}+i\,\hat{E}_{xr}
\end{array}
\right)\quad .
$$

Die Amplitude in $y$-Richtung ist

$$\hat{E}_y=\sqrt{\hat{E}_{yr}^2+\hat{E}_{yi}^2}$$

und die Phase

$$\delta=\arctan\left(\frac{\hat{E}_{yi}}{\hat{E}_{yr}}\right)\quad .$$ 

Die beiden Parameter Neigungswinkel $\alpha$ und Polarisationsellipse $\epsilon$ können aus der Abbildung oben konstruiert werden und sind

$$\tan 2\alpha=\frac{2\hat{E}_x\hat{E}_y\cos\delta}{\hat{E}_x^2-\hat{E}_y^2}$$

bzw.

$$\sin 2\epsilon=\frac{2\hat{E}_x\hat{E}_y\sin\delta}{\hat{E}_x^2-\hat{E}_y^2}$$

<b> Beispiel</b>

Zeigen Sie, dass der Jones Vektor 

$$\frac{1}{14}\left(
\begin{array}{c}
3\\
2+i
\end{array}
\right)$$
elliptisch polarisiertes Licht darstellt.


<b>Lösung</b>

$\delta = \varphi_y-\varphi_x=\arctan (\hat{E}_{yi}/\hat{E}_{yr})=26.6^\circ$

$\hat{E}_x/\hat{E}=3/\sqrt{14}$

$\hat{E}_y/\hat{E}=\sqrt{5}/\sqrt{14}$


Neigungswinkel:

$$\alpha=\frac{1}{2}\arctan\frac{2\frac{\hat{E}_y}{\hat{E}_x}}{1-\left(\frac{\hat{E}_y}{\hat{E}_x}\right)^2}\,\cos\delta=35.8^\circ$$



### 10.4 Matrizendarstellung von Polarisatoren


Optische Elemente, die den Polarisationszustand des transmitterten Lichtes verändern, nennt man Polarisatoren.

<img src="figures/10.4_1.png" alt="10.4_1.png" style="width: 500px;"/>

Wir schauen uns den Linerapolarisator, die Verzögerungsplatte, und die Polarisationsdreher genauer an.

Der <i> Linearpolarisator</i> läßt nur eine Polarisationsrichtung durch. In der Abbildung trifft unpolarisiertes Licht auf den Polarisator und nur die vertikal polarisierte Komponente verläßt den Linearpolaristor.

Die <i>Verzögerungsplatte</i> bewirkt eine Phasenverzögerung zwischen den Komponenten $\vec{E}_x$ und $\vec{E}_y$ von $\delta$. Dies kommt daher zustande, dass die Komponenten eine von der Polarisationsausrichtung abhängige Ausbreitungsgeschwindigkeit haben. Die schnelle Achse (SA) wird normalweise an Verzögerungsplatten angezeigt (s. Abb. unten). Wenn die Phasendifferenz zwischen der langsamen Achse (LA) und der dazu senkrechten schnellen Achse (SA) $\delta=90^\circ$ ist, spricht man von einer Viertelwellenplatte ($\lambda/4$-Platte). Wenn die Phasendifferenz $\delta=180^\circ$ ist, dann von einer Halbwellenplatte ($\lambda/2$-Platte).

Ein <i>Polarisationsdreher</i> dreht die Schwingungsrichtung von linear polarisiertem Licht um einen definierten Winkel. Hier dreht der Polarisationsdreher die vertikale Polarisation um den Winkel $\beta$.

### 10.5 Jones-Matrizen der Polarisatoren

Wir leiten nun Matrizen ab, die die drei oben genannten Typen von Polarisatoren darstellen und auf die Jones Vektoren wirken. 


#### 10.5.1 Linearpolarisator
Wir benutzen eine $2\times 2$ Matrix mit den Elementen $a, b, c, d$ und nehmen das Beispiel eines vertikal ($y$-Richtung) wirkenden Linearpolarisators. Dann ist

$$\left(\begin{array}{cc}
a & b\\
c & d
\end{array}\right)
\left(\begin{array}{c}
0\\
1
\end{array}\right)=
\left(\begin{array}{c}
0\\
1
\end{array}\right)
$$

Diese Gleichung ist gleichbedeutend mit 

\begin{eqnarray*}
a\cdot 0 + b\cdot 1 &=&0\\
c\cdot 0 + d\cdot 1 &=&1
\end{eqnarray*}

und daher ist $b=0$ und $d=1$. Um $a$ und $c$ zu bestimmen nehmen wir eine horizontal polarisierte Welle und können schreiben:

$$\left(\begin{array}{cc}
a & b\\
c & d
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{c}
0\\
0
\end{array}\right)
$$

\begin{eqnarray*}
a\cdot 1 + b\cdot 0 &=&0\\
c\cdot 1 + d\cdot 0 &=&0\quad .
\end{eqnarray*}

Damit ist $a=0$ und $c=0$ und die Jones Matrix eines linearen Polarisators mit vertikaler Polarisationsachse ist:

$$\mathrm{M}=\left(\begin{array}{cc}
0 & 0\\
0 & 1
\end{array}\right)$$

Ganz allgemein, ein Linearpolarisator mit Transmissionsachse unter dem Winkel $\theta$ relativ zur $x$-Achse ist

$$\mathrm{M}=\left(\begin{array}{cc}
\cos^2\theta & \sin\theta\,\cos\theta\\
\sin\theta\,\cos\theta & \sin^2\theta
\end{array}\right)\tag{10.5.1}$$

Herleitung als Übung.

#### 10.5.2 Verzögerungsplatte

Für eine Phasenverzögerung benötigen wir folgende Transformation 

$$\hat{E}_x\,e^{i\,\varphi_x}\quad\mathrm{nach}\quad \hat{E}_x\,e^{i(\varphi_x+\Phi_x)}$$ 

und

$$\hat{E}_y\,e^{i\,\varphi_y}\quad\mathrm{nach}\quad \hat{E}_y\,e^{i(\varphi_y+\Phi_y)}$$

Damit haben wir folgende Matrixgleichung

$$\left(\begin{array}{cc}
e^{i\,\Phi_x} & 0\\
0 & e^{i\,\Phi_y}
\end{array}\right)
\left(\begin{array}{c}
\hat{E}_x\,e^{i\,\varphi_x}\\
\hat{E}_y\,e^{i\,\varphi_y}
\end{array}\right)=
\left(\begin{array}{c}
\hat{E}_x\,e^{i(\varphi_x+\Phi_x)}\\
\hat{E}_y\,e^{i(\varphi_y+\Phi_y)}
\end{array}\right)
$$

Damit ist die Matrix der Phasenverzögerers


$$\mathrm{M}=\left(\begin{array}{cc}
e^{i\,\Phi_x} & 0\\
0 & e^{i\,\Phi_y}
\end{array}\right)\tag{10.5.2}$$


Wir betrachtenn die $\lambda/4$-Platte für die $\delta=\varphi_y-\varphi_x+\Phi_y-\Phi_x=\pm\pi/2$. Für $\delta=+\pi/2$ ist die schnelle Achse vertikal, für $\delta=-\pi/2$ ist diese horizontal. Setzen wir $\Phi_x=0$ und $\Phi_y=+\pi/2$ und $\varphi_x=\varphi_y=0$ dann ist eine (verbreitet Form) der Matrix für die SA vertikal

$$\mathrm{M}=\left(\begin{array}{cc}
e^{0} & 0\\
0 & e^{i\,\pi/2}
\end{array}\right)=\left(\begin{array}{cc}
1 & 0\\
0 & i
\end{array}\right)$$

und für die SA horizontal

$$\mathrm{M}=\left(\begin{array}{cc}
e^{0} & 0\\
0 & e^{-i\,\pi/2}
\end{array}\right)=\left(\begin{array}{cc}
1 & 0\\
0 & -i
\end{array}\right)$$


Bei den $\lambda/2$-Platten $\delta=\pm\pi$ sind die Jones Matrizen für die schnelle und langsame Achse identisch.

Schnelle Achse vertikal:

$$\mathrm{M}=\left(\begin{array}{cc}
e^{0} & 0\\
0 & e^{i\,\pi}
\end{array}\right)=\left(\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right)$$

Schnelle Achse horizontal:

$$\mathrm{M}=\left(\begin{array}{cc}
e^{0} & 0\\
0 & e^{-i\,\pi}
\end{array}\right)=\left(\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right)$$

#### 10.5.3 Polarisationsdreher

Ein Polarisatonsdreher um den Winkel $\beta$ kann man mit einer Rotationsmatrix schreiben

$$\left(\begin{array}{cc}
a & b\\
c & d
\end{array}\right)
\left(\begin{array}{c}
\cos\theta\\
\sin\theta
\end{array}\right)=
\left(\begin{array}{c}
\cos(\theta+\beta)\\
\sin(\theta+\beta)
\end{array}\right)
$$

und nach kurzer Rechnung erhält man


$$\mathrm{M}=\left(\begin{array}{cc}
\cos\beta & -\sin\beta\\
\sin\beta & \cos\beta
\end{array}\right)$$

#### 10.6.2  Polarisation bei Reflexion an dielektrischen Oberflächen

Definition der Einfallsebene:
<img src="figures/10.6_3.png" alt="10.6_2.png" style="width: 550px;"/>
Bild entnommen von [Uni Ulm](http://wwwex.physik.uni-ulm.de/lehre/el-2016/el-2016se40.html) 

Unpolarisiertes Licht kann in zwei Polarisationsrichtungen $E_s$ und $E_p$ aufgeteilt werden. 

<img src="figures/10.6_2.png" alt="10.6_2.png" style="width: 550px;"/>

$E_s$ ist die Polarisationsrichtung senkrecht zur Einfallsebene und damit parallel zur dielektrischen Oberfläche. Influenzierte Dipole strahlen diese Ausrichtung maximal ab. Bei der $E_p$ Richtung ist dem nicht so. Dort hängt die Abstrahlungsintensität $I\propto \sin^2\alpha_d$ von dem Winkel $\alpha_d$ ab. Dieser ist 

$$\alpha_d=90^\circ-(\epsilon+\epsilon')\quad .$$


Wenn $\epsilon+\epsilon'=90^\circ$ ist, wird überhaupt keine Abstrahlung der $E_p$-Komponente ermöglicht. Das heisst, dass von unserem unpolarisiertem Licht nur die $E_s$ Komponente reflektiert wird. Das Licht ist daher polarisiert in der $E_s$-Richtung.

Nach dem Brechungsgesetz ist 

$$n\,\sin\epsilon_p=n'\,\sin\epsilon'$$

Für totale Polarisation in $E_s$-Richtung ist $\epsilon'=90^\circ-\epsilon_p$ und für den Winkel der Reflektion

$$\epsilon_p=\arctan\left(\frac{n'}{n}\right)\quad .$$

Dieser Winkel wird <i>Brewsterwinkel</i> genannt und ist für Luft nach Glas ($n'\approx 1.5$) ca $56^\circ$.

Da nur ein kleiner Teil der Intensität wird reflektiert wird. Deswegen behilft man sich mit einem Stapel von Glasflächen um eine höhere Ausbeute zu erreichen. Die Abbildung unten zeigt einen Polarisator, der auf der Polarisation von Licht unter dem Brewesterwinkel basiert.


<img src="figures/10.6_4.png" alt="10.6_4.png" style="width: 450px;"/>


Ein Brewster Fenster läßt Licht zu 100% durch. Dies ist der Fall wenn das Licht unter dem Brewster-Winkel p-Polarisiert auftritt. Diese erste Reflektion und auch die zweite Reflektion finden jewewils unter dem Brewster Winkel statt. Die zweite Reflektion ist für den Übergang von Glas nach Luft.

<img src="figures/10.6_5.png" alt="10.6_5.png" style="width: 400px;"/>


#### 10.6.3 Polarisation durch Streuung

Die Strahlungscharakteristik eines Herzschen Dipols:

<img src="figures/10.6_6.png" alt="10.6_6.png" style="width: 450px;"/>

Der Dipol strahlt senkrecht zur Schwingungsebene ab. Kleine Streuer mit Durchmesser $d\ll\lambda$, sogenannte Rayleigh Streuung haben ein Intensität im gestreuten Licht von 

$$I_{streu}\propto I_{ein}\,\omega^4\sin^2\alpha\tag{10.2}\quad ,$$

wobei $\omega$ die Kreisfrequenz des Lichtes ist, und $\alpha$ der Winkel zwischen der Dipolachse und der Beobachtungsrichtung. Damit ist die Streuintensität $I_{streu}\propto \lambda^{-4}$.

Da das gestreute elektrische Intensität abhängig von der Orientierung des elektrischen Feldes ist, erhalten wir unterschiedliche Polarisationen in den verschiedenen Richtungen.

<img src="figures/10.6_7.png" alt="10.6_7.png" style="width: 550px;"/>


#### 10.6.4 Polarisation durch Doppelbrechung

In Kristallen, kann der Brechungsindex von der Richtung des einfallenden Lichtes abhängen. Die Achse mit der höchsten Symmetrie nennt sich optische Achse. Ein unpolarisierter Strahl parallel zur optischen Achse in einem optisch isotropen Material sehen beiden Komponenten der Polarisation jeweils einen homogenen Kristall. Daher haben dann auch beiden Polarisationsrichtungen die gleiche Brechzahl. Solch ein Strahl wird ordentlicher Strahl genannt und hat die Brechzahl $n_o$. Tritt der Strahl hingegen senkrecht auf die optische Achse erfahren die beiden Polarisationsrichtungen eine unterschiedliche Kristallsymmetrie und die Brechzahl des sogenanten außerordentlichen Strahles hat die Hauptbrechzahl $n_{ao}$.

Die unten stehende Abbildung zeigt in a) einen Strahl in Richtung der optischen Achse. Hier erfahren beide Polarisationsrichtungen den gleichen Brechungsindex $n_o$. Wenn der Kristall aber wie in b) geschnitten ist, breitet sich die vertikale Polarisation mit dem $n_o$ aus und die horizontale Polarisation mit $n_{ao}$ aus. 
<img src="figures/10.6_9.png" alt="10.6_9.png" style="width: 450px;"/>

Damit kann man Verzögerungsplatten herstellen. Der Gangunterschied der beiden Polarisationsrichtungen ist

$$\Delta=\left|n_{ao}-n{o}\right|\,d\quad .$$

Trifft unpolarisiertes Licht unter beliebigem Winkel auf einen doppelbrechenden Kristall wird das Brechungsgesetz verletzt. Das Beispiel unten zeigt, dass der Strahl obwohl der senkrecht auf den Kristall fällt, wird er in einen ordentlichenn (dem Brechungsgesetz) folgenden Strahl und in einen außerordentilichen Stahl gespalten. 
<img src="figures/10.6_8.png" alt="10.6_8.png" style="width: 550px;"/>

Beide Teilstrahlen sind zueineander polarisiert.

#### 10.6.5 Polarisation durch optische Aktivität

#### 10.6.6 Spannungsoptik
