# 7. Interferometrie

## 7.1 Michelson Interferometer

Das [Michelson-Morley-Experiment](https://de.wikipedia.org/wiki/Michelson-Morley-Experiment) hat die moderne Physik mit begründet. Michelson hat zur Messung der [Äthergeschwindigkeit](https://link.springer.com/article/10.1007%2FBF01528662) ein Aperatur aufgebaut, die es ermoöglicht die Lichtgeschwindigkeit in verschiedene Richtungen zu messen. Diesr Michelson Interferometer wird seitdem für Präzissionsmessungen genutzt. Der letzte große Erfolg des Michelson Interferometer war die Detektion von Gravitationswellen. 

Der Aufbau ist unten skizziert: Ein Strahl (1) von der Lichtquelle wird an der Obereite eines halbdurchlässigen Spiegels $ST$ reflektiert, ein Strahl wird um $90^\circ$ umgelenkt (2) und ein zweiter Strahl (3) geht durch den halbdurchlâssigen Spiegel geradeaus. Die Amplitde der beiden Strahlen ist ungefähr gleich groß.

Die beiden Strahlen (2) und (3) werden mit den Spiegelns $S_2$ und $S_1$ zurückgeworfen, Die Hälfte des Strahls (2) läuft durch den halbdurchlässigen Spiegel durch, und auch die Hälfte des Strahls (3) wird reflekiert. Beide Teilstrahlen ergebn den Strahl (4) und intereferieren auf dem Detektor.

In Experiment wird häufig nicht ein dünner Strahl genommen, sondern ein mit einer negativen Linse aufgeweiteter Laserstrahl.  

<img src="figures/7.1_1.png" alt="7.1_1.jpg" style="width: 450px;"/>


Obwohl auf den ersten Blick ähnlich, gibt es zwei wichtige Unterschied zwischen den Strahlen 2 und 3. Strahl 2 wird an der Vorderseite des Strahlteilers einmal mit Phasensprung $\lambda/2$ reflektiert und läuft nach Reflektion durch den Strahlteiler. Strahl 3 läuft dreimal durch den Strahlteiler und wird ohne Phasensprung am Strahlteiler reflektiert. Beide Strahlen (2 und 3) erfahren einen zweiten Phasensprung an den Spiegeln $S_2$ und $S_1$. Zur Kompensation des längeren optischen Weges von Strahl 3 kann eine Glasplatte in den Strahlengang 2 eingebaut werden. Der zusätzliche Phasensprung von Strahl 2 muss aber dennoch berücksichtigt werden um auf den Gangunterschied $\Delta$ zu kommen:

$$\Delta=\Delta_w+\frac{\lambda}{2}=2\,d+ \frac{\lambda}{2}\quad .$$

Die Strecke $d$ ist die Verschiebung des Spiegels senkrecht zur optischen Achse/Strahles.

Wenn beide Strahlen 1 und 2 gleich hell sind nehmen wir unsere Gleichung für das Interferenzmuster als Funktion des Phasendifferenz $\delta$:

$$I=4\,I_0\cos^2\left(\frac{\delta}{2}\right)\quad.$$

Der Zusammenhang zwischen Phasendifferenz und Gangunterschied ist:

$$\delta=k\,\Delta=2\pi\frac{\Delta}{\lambda}$$


Für dunkle Streifen ist $\Delta=2\,d+\lambda/2=(2m+1)\lambda/2$ oder 

$$2\,d=m\,\lambda$$

Eine Verschiebung eines Spiegels von $d=\lambda/2$ erzeugt einen dunkel-hell-dunkel Übergang auf dem Streifensystem. 

Wie bereits oben angedeutet, wird der Laserstrahl typischerweise aufgeweitet. Dadurch entsteht ein Ringsystem. Hier könnte man den Detektor auf das Zentrum ausrichten und die Anzahl der Übergänge von dunkel auf dunkel messen. Jeder solche Übergang entspricht einer Weglängenverschiebung $d=\lambda/2$. 

## 7.2 Zeitliche Kohärenz

Der Begriff der Kohärenz wurde eingeführt um die Eigenschaft, das Licht Interferenz erzeugen kann, zu quantifizieren. Im Laufe wurde der Begriff präzisiert und beschreibt heute die Korrelation von zwei optischen Größen.

Schauen wir uns zur Definition der zeitlichen Kohärenz das Michelson Inteferometer an:

<img src="figures/7.2_1.png" alt="7.2_1.jpg" style="width: 450px;"/>

Die Verschiebung des Spiegels um den Abstand $d$ führt zu einer zeitlichen Verzögerung eines der beiden Lichtstrahle. Nennen wir den Stral der auf den verschiebaren Spiegel fällt $E_2$ und die Verschiebung $d$ positiv zu längeren Lauflängen besteht die Beziehung zwischen dem vertikalen Strahl $E_1$ und dem horizontalen Strahl $E_2$

$$E_2(t)=E_1(t+\tau)\quad\mathrm{oder}\quad E_1(t)=E_2(t-\tau)$$

Das elektrische Feld auf dem Schirm ist dann

$$E(t)=E_1(t)+E_2(t)=E_1(t)+E_1(t+\tau)$$

Wir messen die Intensität und diese ergibt sich aus dem Realteil des Betragsquarats der elektrischen Feldstärke. Beachten Sie, dass die $<\,>$-Klammern einen zeitlichen Mittelwert beschreiben. 

\begin{eqnarray*}
I&=&\left<E\,E^*\right>=\left<\left(E_1+E_2\right)\left(E_1+E_2\right)^*\right>\\
&=&\left<E_1\,E_1^*\right>+\left<E_2\,E_2^*\right>+\left<E_2\,E_1^*\right>+\left<E_1\,E_2^*\right>\\
&=& I_1+ I_2+ 2\, Re \left\{ \left< E_1^* E_2 \right> \right\}\\
&=& 2 I_1+ 2\, Re \left\{ \left< E_1^* E_2 \right> \right\}\tag{7.2.1}
\end{eqnarray*}


In Gleichung (7.2.1) haben wir angenommen, dass beide elektrischen Felder gleiche Amplituden haben. Nun berechnen wir den rechten Term in Gl. (7.2.1) aus. Dazu führen wir gleich die Selbstkohärenzfunktion $\Gamma$ ein und benutzten $E_2=E_1(t+\tau)$:

\begin{eqnarray*}
\Gamma(\tau)&=&\left<E_1(t)^*\,E_1(t+\tau)\right>\\
&=&\lim_{T_m\rightarrow\infty}\frac{1}{T_m}\int_{-T_m/2}^{+T_m/2}E_1(t)^*\,E_1(t+\tau)\,\mathrm{d}t\\
&=&\lim_{T_m\rightarrow\infty}\frac{1}{T_m}\int_{-T_m/2}^{+T_m/2}\left|E_0\right|^2\,e^{i\omega t}\,e^{-i\omega\left(t+\tau\right)}\,\mathrm{d}t\\
&=&\left|E_0\right|^2\,e^{-i\omega \tau}=I_1\,e^{-i\omega\tau}
\end{eqnarray*}

Damit wird aus Gl. (7.2.1)

\begin{eqnarray*}
I(\tau)&=&2\,I_1+2\, Re \left\{ \Gamma(\tau) \right\}\tag{7.2.2a}\\
&=&2\,I_1+2\,I_1\,Re\left\{ e^{-i\omega \tau} \right\}\\
&=&2\,I_1+2\,I_1\,\cos(\omega\tau)=2I_1\left(1+\cos(\omega\tau)\right)\tag{7.2.2}
\end{eqnarray*}

<img src="figures/7.2_2.png" alt="7.2_2.jpg" style="width: 450px;"/>

Wir können die komplexe Selbstkohärenzfunktion normieren

$$\gamma(\tau)=\frac{\Gamma(\tau)}{\Gamma(0)}\tag{7.2.3}$$

$\gamma$ wird Kohärenzgrad genannt. Da $\Gamma(0)=I_1$ eine relle Zahl ist und den größte Wert annimmt erhalten wir

$$\left|\gamma(\tau)\right|\leq 1\quad .$$

Damit ergibt sich für die Intensität $I(\tau)$ 

\begin{eqnarray*}
I(\tau)&=&2\,I_1+2\,I_1\, Re \left\{ \gamma(\tau) \right\}\\
&=&2\,I_1\left(1+\,Re\left\{\gamma(\tau)\right\}\right)\quad .
\end{eqnarray*}

Der Kontrast der Interferenzstreifen ergibt sich aus der maximalen und minimalen Intensität. Die zeitlichen Verschiebungen seinen $\tau_1$ fuür die maximale und $\tau_2$ für die minimale Intensität. Im Interval $[\tau_2-\tau_2)$ gibt es nur ein Maximum und ein Minimum, dann ist der Kontrast in diesem Interval  

$$K(\tau)=\frac{I_{max}(\tau_1)-I_{min}(\tau_2)}{I_{max}(\tau_1)+I_{min}(\tau_2)}\quad .$$

Aus Gleichung (7.2.2a) folgt die maximale Intensität wird beim Maximum von $Re\left\{\Gamma(\tau)\right\}$ bei $\tau_1$ und die minimale Intensität beim Minimum von $Re\left\{\Gamma(\tau)\right\}$ bei $\tau=\tau_2$ erreicht. Weiterhin ist der Betrag von $\Gamma(\tau)$ im Interval annähernd konstant und somit

$$Re\left\{\Gamma(\tau_1)\right\}=\left|\Gamma(\tau)\right|\quad \mathrm{and}\quad Re\left\{\Gamma(\tau_2)\right\}=-\left|\Gamma(\tau)\right|$$
Damit sind die Intensitäten

\begin{eqnarray*}
I_{max}(\tau_1)&=&2I_1+2\,Re\left\{\Gamma(\tau_1)\right\}=2I_1+2\left|\Gamma(\tau)\right|\quad \mathrm{and}\\
I_{min}(\tau_2)&=&2I_1+2\,Re\left\{\Gamma(\tau_2)\right\}=2I_1-2\left|\Gamma(\tau)\right|
\end{eqnarray*}

und die Kontrastfunktion vereinfacht sich

\begin{eqnarray*}
K(\tau)&=&\frac{2I_1+2|\Gamma(\tau)|-2I_1+2|\Gamma(\tau)|}{2I_1+2|\Gamma(\tau)|+2I_1-2|\Gamma(\tau)|}\\
&=&\frac{4|\Gamma(\tau)|}{4I_1}=\frac{|\Gamma(\tau)|}{I_1}=\frac{|\Gamma(\tau)|}{\Gamma(0)}\\
&=&|\gamma(\tau)|\tag{7.2.4}
\end{eqnarray*}

Damit ist die Kontrast gleich dem komplexen Kohärenzgrad.

Bei Quasi monochromatischem Licht deren *Selbstkohärenz* in den Ursprung hinein spiralisiert nimmt der Kontrast monoton ab.

<img src="figures/7.2_3.png" alt="7.2_3.jpg" style="width: 450px;"/>

Für eine harmonische Welle ist

$$\gamma(\tau)=e^{-i\omega \tau}$$

Damit ist der Kontrast konstant 1:

$$K(\tau)=|\gamma(\tau)|=1.$$

<img src="figures/7.2_4.png" alt="7.2_4.jpg" style="width: 450px;"/>

Eine harmonische Welle behält ihre Interferenzeigenschaften bei beliebigen Phasenverschiebungen. Solches Licht nennt man vollständig kohärent. Dies ist ein Grenzfall und kann approximativ mit einem stabilisierten Einzelmodenlaser erreicht werden.

Der zweite Grenzfall ist vollständig inkohärentes Licht für das $|\gamma(\tau)|=0$ wenn $\tau\ne0$ ist:

<img src="figures/7.2_5.png" alt="7.2_5.jpg" style="width: 450px;"/>

Solches Licht besteht aus vielen Wellenlängen mit statistisch verteilten Phasen und Beispiele für fast vollständig inkohärentes Licht sind Glühlampen und Tageslicht.

Licht zwischem vollständig kohärent und vollständig inkohärent wird teilkohärent gennannt mit $0\leq|\gamma(\tau)|\leq 1$.

Viele natürliche und künstlichen Lichquellen haben einen monoton abfallenden Kontrast. Um deren Kohärenz zu beschreiben wird die Kohärenzzeit $\tau_c$ eingeführt. Dies ist die Zeit, bei der die Kontrastfunktion auf $1/e$ abgefallen ist. Die entsprechende Länge ist die Kohärenzlänge 

$$l_c=c\,\tau_c\quad .\tag{7.2.5}$$

Typische Längen sind Mikrometer für Glühlampen und Kilometer für Einzelmoden Laserlicht.

<img src="figures/7.2_6.png" alt="7.2_6.jpg" style="width: 450px;"/>

Die Kohärenz muss keine monoton fallende Funktion sein. Zum Beispiel wenn wir die Überlagerung von zwei harmonischen Wellen mit unterschiedlicher Frequenz und gleicher Amplitude betrachten, erhalten wir einen Schwebung. Wie schaut der Kontrast so einer Lichtquelle aus, wie sie z.B. in einem Zwei-Moden Laser vorkommt.

$$E(t)=E_0\,e^{-i\omega_1 t}+E_0\,e^{-i\omega_2 t}$$

Dann ist die Selbstkohärenz 

\begin{eqnarray*}
\Gamma(\tau)&=&\lim_{T_m\rightarrow\infty}\frac{1}{T_m}\int_{-T_m/2}^{+T_m/2}\left[E_0^* e^{i\omega_1 t}+E_0^* e^{i\omega_2 t}\right]\cdot\left(E_0 e^{-i\omega_1(t+\tau)}+E_0 e^{-i\omega_2(t+\tau)}\right)\,\mathrm{d}t\\
&=&\lim_{T_m\rightarrow\infty}\frac{|E_0|^2}{T_m}\int_{-T_m/2}^{+T_m/2}\left(e^{-i\omega_1\tau}+e^{-i\omega_2\tau}+\right.\\
&&\underbrace{\left. e^{-i\omega_1 \tau} e^{-i(\omega_1-\omega_2)t}+
e^{-i\omega_2 \tau} e^{-i(\omega_2-\omega_1)t}\right)}_{\mbox{Mittelwert ist 0}}\,\mathrm{d}t\\
&=&|E_0|^2\left(e^{-i\omega_1 \tau}+e^{-i\omega_2 \tau}\right)\quad.
\end{eqnarray*}

Mit Gl. (7.2.3) und $\Gamma(0)=2|E_0|^2$ erhalten wir für die normierte Selbstkohärenzfunktion

$$\gamma(\tau)=\frac{1}{2}\left(e^{-i\omega_1\tau}+e^{-i\omega_2\tau}\right)$$

Den Kontrast berechnen wir direkt

\begin{eqnarray*}
K(\tau)&=&|\gamma(\tau)|=\frac{1}{2}\left|e^{-i\omega_1\tau}+e^{-i\omega_2\tau}\right|\\
&=&\frac{1}{2}\sqrt{\left(e^{-i\omega_1\tau}+e^{-i\omega_2\tau}\right)\left(e^{i\omega_1\tau}+e^{i\omega_2\tau}\right)}\\
&=&\frac{1}{2}\sqrt{2+2\cos\left(\omega_1-\omega_2\right)\tau}=\frac{1}{2}\sqrt{4\cos^2\frac{\left(\omega_1-\omega_2\right)}{2}\tau}\\
&=&\left|\cos\left(\frac{\omega_1-\omega_2}{2}\right)\right|\tag{7.2.6}
\end{eqnarray*}


<img src="figures/7.2_7.png" alt="7.2_7.jpg" style="width: 450px;"/>

In diesem Fall ist die Kontrastfunktion $K(\tau)$ periodisch mit der Verschiebung $\tau$ und wir können keine eindeutige Kohärenzzeit oder Kohärenzlänge bestimmen. Eine Möglichkeit das zu umgehen, ist das erste Minimum als Kohärenzlänge/Zeit zu nehmen.

## 7.3 Vielstrahlinterferenz am Fabry Perot

Das Fabry Perot Interferometer besteht aus 2 teildurchlässigen Spiegel, die parallel aufgestelly sind. Einer der beiden Spiegel lässt sich verschieben. Strahlen im Interferometer werden vielfach zwischen den Spiegeln hin und her reflektiert. Wir betrachten das sich ergebenede Intereferenzmuster das mit einer Linse auf einen Schirm projektiert wird. 
Zur Berechnung der Intereferenz wird nur die räumliche Phase berücksichtigt, d.h. $E=E(z)$.

<img src="figures/7.3_1.png" alt="7.3_1.jpg" style="width: 450px;"/>

Die einfallende Welle ist 

$$E_e=E_0\,e^{ikz}$$

und wir führen die reellen Reflektions- und Transmissionskoeffizienten ein

$$r=\frac{E_r}{E_e}\quad\mathrm{und}\quad t=\frac{E_t}{E_e}\quad .$$

Weiterhin nehmen wir an, dass die Spiegel das Licht nicht absorbieren. Daher gilt die Energieerhaltung

$$r^2+t^2=1$$

Der Strahl, der durch den ersten Spiegel geht hat

$$E_t=t\,E_e(z=0)=t\,E_{e0}$$

nachdem der Strahl durch den zweiten Spiegel geht

$$E_1=t\,E_t\,e^{ikL}=E_{e0}\,t^2\,e^{ikL}$$

Beim Durchgang durch das Spiegelpaar wird die Phase um die Wegläge $kL$ vergrößert.

Betrachten wir nun den am Spiegel 2 reflektierten Strahl:

$$E_{1r}=t\,r\,E_{e0}e^{ikL}$$

diser Strahl wird am ersten Spiegel reflektiert und am zweiten Spiegel transmittiert

$$E_2=t^2\,r^2\,E_{e0}^{ik2L}\quad ,$$

also zwei mal transmittiert und zwei mal reflektiert. Allgemein ergibt sich für jeden weiteren Strahl

$$E_n=E_{n-1}\,r^2\,e^{ik2L}\quad n=2,3,\dots $$

oder 

$$E_n=E_{e0}\,t^2\,e^{ikL}\,r^{2(n-1)}\,e^{ik2(n-1)L}\quad .$$

<img src="figures/7.3_2.png" alt="7.3_2.png" style="width: 500px;"/>


Die ebene Wellenfront hinter dem Fabry Perot ist die Überlagerung aller Wellenfronten $E_1,E_2,E_3,\dots$ :

\begin{eqnarray*}
E_t&=&\sum_{n=1}^{\infty}E_n=E_{e0}\,t^2\,e^{ikL}\,\sum_{n=1}^{\infty}r^{2(n-1)}\,e^{ik2(n-1)L}\\
&=&E_{e0}\,t^2\,e^{ikL}\,\sum_{n=0}^{\infty}r^{2n}\,e^{ik2nL}\\
&=&E_{e0}\,t^2\,e^{ikL}\,\sum_{n=0}^{\infty}\left(r^2\,e^{iknL}\right)^n\tag{7.3.1}
\end{eqnarray*}

Die Summe einer geometrischen Reihe ist

$$\sum_{n=0}^{\infty}q^n=\frac{1}{1-q}\quad \mathrm{für}\quad |q|<1.$$

Setzen wir jetzt $q=r^2\,e^{ik2L}$ dann ist die Amplitude der Welle hinter dem Fabry Perot

$$E_t=E_{e0}t^2\,e^{ikL}\,\frac{1}{1-r^2\,e^{ik2L}}\quad .$$

Die Intensität lässt sich dann auch berechnen. Zur einfacheren Darstellung wird die Phase $\delta=k2L$  benutzt:

\begin{eqnarray*}
I_t=E_t\,E_t^*&=&E_{e0}\,E_{e0}^*\,t^4\frac{1}{\left(1-r^2\,e^{i\delta}\right)\left(1-r^2\,e^{-i\delta}\right)}\\
&=&I_e\,t^4\,\frac{1}{1+r^4-r^2\,e^{-i\delta}-r^2\,e^{i\delta}}\\
&=&I_e\,t^4\,\frac{1}{1+r^4-2r^2\cos\delta}\quad ,\tag{7.3.2}
\end{eqnarray*}
wobei $I_e=E_{e0}E_{e0}^*$ die Intensität des einfallenden Lichtes ist. Wir vereinfachen Gl. (7.3.2) weiter ($\cos\delta=1-2\sin^2{\delta/2}$).

\begin{eqnarray*}I_t&=&I_e\,t^4\frac{1}{1+r^4-2r^2\left(1-2\sin^2(\delta/2)\right)}\\
&=&I_e\,t^4\frac{1}{(1-r^2)^2+4r^2\sin^2(\delta/2)}\\
&=&I_e\,t^4\frac{1}{1+\left(\frac{2r}{1-r^2}\right)^2\sin^2(\delta/2)}\\
&=&\frac{I_e}{1+K\sin^2(\delta/2)}=\frac{I_e}{1+K\sin^2 kL}
\end{eqnarray*}
Hier haben wir $r^2+t^2=1$ ausgenutzt und die Koeffizient der Finesse 

$$F=\left(\frac{2r}{1-r^2}\right)^2$$ eingeführt. 

Die Abbildung unten zeigt die normierte Transmission des Fabry Perot

$$T_I=\frac{I_t}{I_e}=\frac{1}{1+K\sin^2 k L}$$

<img src="figures/7.3_3.png" alt="7.3_3.png" style="width: 500px;"/>

Diese Funktion heißt Airy Funktion. Sie ist periodisch in $L$ und $k$. Wenn wir den Abstand $L$ konstant halten und die Wellenzahl $k=2\pi/\lambda$ oder die Wellenlänge $\lambda$ ändern, ändert sich die Intensität der Transmission. Insbesonders wird der Zähler periodisch und damit erhalten wir komplette Transmission für gewisse Wellenlängen. D.h. das Fabry Perot ist ein optischer Filter (Kammfilter).

Ein weiterer interessanter Aspekt ist, dass obwohl ein Spiegel sagen wir 99.9% reflektiert und daher nur 0.1% des Lichtes durchlässt, in der Fabry Perot Anordnung 100% der selbe Spiegel 100% durchlässt.

<img src="figures/7.3_4.png" alt="7.3_4.png" style="width: 500px;"/>

Die Reflektivität eines Spiegels wird also nicht alleine vom Spiegel bestimmt, sondern auch vom Aufbau. 

Mit Erhöhung der Reflektivität $r$ werden die Transmissionpeaks schmaler. Der Abstand der Peaks ist von großer Wichtigkeit. 

Transmission wird erreicht wenn

$$n\pi=k_n\,L=\frac{2\pi}{\lambda_n}L \quad \mathrm{oder}\quad n\frac{\lambda_n}{2}=L\quad n=0,1,2,\dots$$ 

Das heisst aber auch, dass nicht die absolute Wellenlänge gemessen werden kann sonder nur Wellenlängen relativ zu den Peaks. Der Abstand wird gewöhnlich als Differenz der Frequenz $\Delta \nu$ angegeben und nennt sich der freie Spektralbereich (free spectral range im Englischen, oder auch FSR). Er kann folgendermaßen bestimmt werden:

$$\pi=(k_{n+1}-k_n)L=\Delta k\, L=\frac{2\pi\Delta \nu}{c}L$$

Hier haben wir auch den Zusammenhang $\Delta k=\Delta\nu\,2\pi/c$ ausgenutzt.

Der freie Spektralbereich ist daher 

$$\Delta \nu=\frac{c}{2L}$$

und zwei Frequenzen die weniger als $\Delta \nu$ von einander entfernt sind, können noch aufgelöst werden. Die absolute Frequenz/Wellenlänge muß andersweitig gemessenen werden.

Weiterhin gibt es neben der FSR die Finesse. Diese gibt an, wie viele Spektrallinien innerhalb der FSR gemessen werden können. 

$$F=\frac{\Delta \nu}{\nu_h}$$ 

Hier ist $\nu_h$ die Halbwertsbreite der Airy Funktion. Nach kurzer Rechnung kann man zeigen, dass 

$$F=\frac{\pi r}{1-r^2}$$

ist.

### 7.4.1 Differenz-Frequenz Messung

Nehmen wir an, der Laser verstärkt 2 Moden mit Ferquenzen $\omega_1$ $\omega_2$

$$E(t)=e^{-i\omega_1\,t}+e^{-i\omega_2\,t}$$

Dann ist die Spannung an der Photodiode proportional zur Intentsität

$$U(t)\propto I=2+2\,\cos\left[\left(\omega_1-\omega_2\right)t\right]$$

<img src="figures/7.4_2.png" alt="7.4_2.png" style="width: 400px;"/>
