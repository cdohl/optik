## 4. Abbildungsfehler

Bisher haben wir die paraxiale Näherung für die Strahlenbeschreibung der optischen Abbildung benutzt. Dazu wurden die trigonometrischen Funktionen in eine Taylorreihe (um den Winkel) entwickelt und nur der erste Term der Reihe benutzt. Jetzt werden wir höhre Terme der Reihe benutzten und dabei auch sehen, dass die Abbildung vom Abstand von der optischen Achse abhängt.

\begin{eqnarray}
\sin \epsilon  &=& \epsilon - \frac{\epsilon^3}{3!} + \frac{\epsilon^5}{5!} - \dots \approx \epsilon\\
\cos \epsilon  &=& 1 -  \frac{\epsilon^2}{2!} + \frac{\epsilon^4}{4!} - \dots \approx 1
\end{eqnarray}

### 4.1. Longintudinale und transversale Aberration

Die sphärische Aberration ist der Bildfehler der dadurch entsteht, dass die Strahlen durch eine sphärisch gekrümmte Fläche gebrochen werden. Die Wellenfront ist *nicht* eine sphärisch gekrümmte Wellenfront. Dadurch gibt es keinen eindeutige Fokussierung der Wellenfront. Nur wenn die Strahlen achsnah (paraxial) sind, müssen wir den Abstand $h$ nicht berücksichtigen. Die Abbildung zeigt zwei Strahlen, einmal von der idealen *sphärischen* Wellenfront und eine von einer nichtidealen oder realen Wellenfront. Die ideale Wellenfront fokussiert die Strahlen am Punkt $O'$, die nichtideale Wellenfront fokussiert *nicht* auf einen Punkt. Deren Fokussierung hängt vom Abstand $h$ ab. Um dies mathematisch zu beschreiben, führen wir Strahlen ein, die senkrecht zu den beiden Wellenfront liegen und auf der Höhe $h$ liegen.   

<img src="figures/4.1_1.png" alt="4.1_1.png" style="width: 600px;"/>

Der Abstand der beiden Strahlen beim Schneiden der optischen Achse nennen wir <b> Längsabweichung</b> oder <b>longitudinale Aberration</b>, $\Delta z$. Es gibt auch einen vertikalen Abstand der Strahlen, nämlich in der paraxialen Bildebene (Punkt $O'$) haben die Strahlen eine <b>Querabweichung</b> oder <b>transversale Aberration</b>, $\Delta y$. Wir beschreiben zuerst allgemein in wieweit sich eine optischer Wegdifferenz zwischen den beiden Strahlen $q$ zu den Aberrationen $\Delta z$ und $\Delta y$ führt. Bitte bedenken Sie, dass $q$ eine Funktion von der Höhe $h$ ist. Der Winkel zwischen den beiden Strahlen ist $\alpha$. Im Ausschnitt in obiger Abbildung ist die Entstehung des Abstandes $q(h)$ vergrößert dargestellt. Wir erhalten für den optischen Weg ($n'$ mal dem geometrischen Weg)
 
$$\mathrm{d}q=n'\,\tan\alpha\,\mathrm{d}h\approx n'\,\alpha\,\mathrm{d}h$$

und damit

$$\alpha = \frac{1}{n'}\frac{\mathrm{d}q}{\mathrm{d}h}\quad.\tag{4.1.1}$$

Weiterhin ist über das Bogenmaß 

$$\Delta y= \alpha\,s'\quad,$$

in das wir den Winkel Gl.(4.1.1) einsetzen können

$$\Delta y = \frac{s'}{n'}\frac{\mathrm{d}q}{\mathrm{d}h}\tag{4.1.2}$$

Mit Hilfe des Winkels $\sigma$ berechnen wir die Längsabweichung

$$\tan{\sigma}=\frac{\Delta y}{\Delta z}=\frac{h}{s'}$$

Nach $\Delta z$ aufgelöst finden wir für die longitudinale Aberration

$$\Delta z = \frac{\Delta y\,s'}{h}\quad . \tag{4.1.3}$$

Gleichungen (4.1.2) und (4.1.3) geben die Aberration in Abhängigkeit von $q(h)$ an. Wir müssen nun die optische Weglängenabweichung $q(h)$ für eine gegebene brechende Fläche berechenen um die beiden Aberrationen zu bestimmen. Die Rechnung können schnell aufwändig werden. Selbst für eine einzelne sphärische brechende Fläche müssen wir diverse Näherungen einführen, um auf ein analytisches Ergebnis zu kommmen. Dies machen wir im nächsten Abschnitt.

### 4.2 Sphärische Aberration durch eine Brechung an einer Kugelfläche

<img src="figures/4.2_1.png" alt="4.2_1.png" style="width: 600px;"/>

Die Abbildung oben zeigt Strahlen vom Punkt $O$, die durch eine Kugelfläche gebrochen werden und am Punkt $O'$ ankommen. In der paraxialen Näherung haben alle Strahlen die vom Punkt $O$ auslaufen und im Punkt $O'$ ankommen den gleichen optischen Weg. Diese Näherung lassen wir nun fallen und berechnen den Wegunterschied zwischen dem idealen paraxialen Strahl $OSO'$ und jedem anderen Strahl, der an der sphärisch gekrümmten Fläche gebrochen wird. Diese anderen Strahlen laufen über den Punkt $Q$ auf der Fläche. Der optische Wegunterschied beider Strahlen ist $q(Q)$:

$$q(Q)=\left(OQO'-OSO'\right)_{\mathrm{OW}}\quad ,$$

wobei $\mathrm{OW}$ für optischen Weg steht. In den nachfolgenden Rechnungen müssen wir sehr vorsichtig mit den Vorzeichen sein, und uns an die Konventionen halten (nach links negativ).

Damit ist 
$$q(Q)=\left(nl+n'l'\right)+ns-n's'\quad .\tag{4.2.0}$$

Die beiden Strecken $l$ und $l'$ bestimmen wir mit dem Kosinussatz und dem Winkel $\varphi$:

$$l^2=r^2 + (r-s)^2 -2r(r-s)\cos \varphi\tag{4.2.1}$$
$$l'^2=r^2 + (r-s')^2 -2r(r-s')\cos \varphi\tag{4.2.2}$$

Der Winkel $\varphi$ wird durch die Tayloreihe angenähert

$$\cos \varphi = 1 -\frac{\varphi^2}{2!}+\frac{\varphi^4}{4!}-\dots$$
mit 

$$\varphi\approx \frac{h}{r}$$

und damit

$$\cos \varphi \approx 1-\frac{h^2}{2\,r^2}+\frac{h^4}{24\,r^4}\tag{4.2.3}$$

Setzen wir nun Gl.(4.2.3) in die Gln.(4.2.1) und (4.2.2) ein, erhalten wir nach kurzer kurzer Rechnung:

$$l=-s\left[1+\overbrace{\left(\frac{h^2(r-s)}{rs^2}-\frac{h^4(r-s)}{12r^3s^2}\right)}^{x}\right]^{1/2}$$
$$l'=s\left[1+\left(\frac{h^2(r-s')}{rs'^2}-\frac{h^4(r-s')}{12r^3s'^2}\right)\right]^{1/2}$$

Die beiden Gleichungen sind von der Art $(1+x)^{1/2}$, was wiederum in eine Reihe entwickelt werden kann:

$$\left(1+x\right)^{1/2}=1+\frac{x}{2}-\frac{x^2}{8}+\dots$$

Wenn wir nun auch noch Terme $h^6$ und $h^8$ vernachlässigen ergibt sich als Näherungen für die Gleichungen (4.2.1) und (4.2.2)

$$l=-s\left(1+\frac{h^2(r-s)}{2rs^2}-\frac{h^4(r-s)}{24r^3s^2}-\frac{h^4(r-s)^2}{8r^2s^4}\right)\tag{4.2.4}$$

$$l'=s'\left(1+\frac{h^2(r-s')}{2rs'^2}-\frac{h^4(r-s')}{24r^3s'^2}-\frac{h^4(r-s')^2}{8r^2s'^4}\right)\tag{4.2.5}$$

Setzen wir nun diese Gleichungen in die optischen Wegunterschied Gl.(4.2.0) ein ergibt sich nach kurzer aber intensiver Rechnung

$$q(Q)=\frac{h^2}{2}\underbrace{\left[\left(\frac{n'}{s'}-\frac{n}{s}\right)-\left(\frac{n'-n}{r}\right)\right]}_{\mathrm{Paraxiale\, Näherung = 0}}-
\frac{h^4}{8}\left[-\frac{n}{s}\left(\frac{1}{r}-\frac{1}{s}\right)^2+\frac{n'}{s'}\left(\frac{1}{s'}-\frac{1}{r}\right)^2\right]$$

Der erste Term ist die paraxiale Näherung für eine kugelförmige brechende Fläche und nach dem Fermatschen Prinzip gleich Null. Damit vereinfacht sich die optische Weglänge $q(h)=g\,h^4$ mit

$$g=-\frac{1}{8}\left[\frac{n}{s}\left(\frac{1}{r}+\frac{1}{s}\right)^2+\frac{n'}{s'}\left(\frac{1}{s'}-\frac{1}{r}\right)^2\right]\tag{4.2.6}$$

Die Größe der sphärischen Aberration hängt von $\frac{\mathrm{d}q}{\mathrm{d}h}$ ab, und damit ist diese proportional $h^3$. Die sphärische Aberration ist ein Abbildungsfehler der 3. Ordnung.

#### Beispiel

<img src="figures/4.2_2.png" alt="4.2_2.png" style="width: 600px;"/>

Bestimmen Sie für das parallele Strahlenbündel die sphärischen Längs- und Querabweichungen!

Für die Bestimmung der Strecke $s'$ benutzen wir das Ergebnis aus der paraxialen Näherung

$$\frac{n'}{s'}-\frac{n}{s}=\frac{n'-n}{r}$$

mit $s=-\infty$ finden wir für

$$s'=\frac{n'}{n'-n}\,r\approx 10.7\,\mathrm{cm}$$

Genauso vereinfacht sich Gl.(4.2.6) zu


$$g=-\frac{1}{8}\left[\frac{n'}{s'}\left(\frac{1}{s'}-\frac{1}{r}\right)^{2}\right]$$

und wir haben für

$$\frac{\mathrm{d}q}{\mathrm{d}h}=4h^3\,g$$

Die Querabweichung ist dann 

$$\Delta y = \frac{s'}{n'}\frac{\mathrm{d}q}{\mathrm{d}h}\approx 0.012\,\mathrm{cm}$$

und die Längsabweichung

$$\Delta z = \frac{s'\,\Delta y}{h}=-0.13\,\mathrm{cm}\quad.$$
